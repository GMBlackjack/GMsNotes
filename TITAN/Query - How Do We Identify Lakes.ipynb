## How do we Identify Titan's Lakes?

### Let me count the ways...

First of all, there are a couple of smoking guns.

1) Specular Reflection

Stephan et al. 2010

https://agupubs.onlinelibrary.wiley.com/doi/full/10.1029/2009GL042312





This is the one I work with, direct specular reflections. The light reflects off the liquid's surface directly and slams into the detector, overloading everything. Pretty obvious.

2) Radar Pings

Mastrogiuseppe et al. 2016

https://iris.uniroma1.it/retrieve/e3835316-51ec-15e8-e053-a505fe0a3de9/Mastrogiuseppe_Radar-sounding_2016.pdf



Another pretty obvious one is where you send a radar ping down into an unknown material, and you get two pings back. That's generally indicative of liquid. (This paper's not a discovery as we already knew about the lakes at this point.)

Beyond these two, there are other methods, and in fact we had to start with them first. The paper that first discovered the lakes (which I don't have full access too for some reason...) outlines its methods in the abstract.

Stofan et al. 2007

https://www.nature.com/articles/nature05438

3) Morphology

This is the one that's the most prone to arguments, but sometimes you can just tell something's a lake by it's shape. Or the fact that there are rivers next to it or channels... for the very large seas this is obvious. For smaller ones, ehhhhh, is it filled with liquid or dry...? 

This probably also encompasses the general "flatness" criterion. The lake should appear roughly the same everywhere. Then again, so do the blandlands, so...

4) Radar Darkness

Okay so it's time to figure out what "radar-dark" MEANS. Some digging around revealed what it actually was: radar dark is SMOOTHNESS. Now I have to figure out how radar works and why it gets dark on smooth objects...

...Okay so https://natural-resources.canada.ca/maps-tools-and-publications/satellite-imagery-elevation-data-and-air-photos/tutorial-fundamentals-remote-sensing/microwave-remote-sensing/target-interaction-and-image-appearance/9311 Canada to the rescue. Smooth objects tend to specularly reflect radar, and thus little to no energy comes back as it's generally pushed elsewhere. (This is presumably why the image for Radar Pings has a direct normal incidence). I see...

Naturally Radar Darkness only helps when you have somehting large enough to appear dark to radar, so really small locations aren't helpful. Also particuarly shallow lakes won't help as the subsurface will send a radar ping back, and if it's below the instrument's resolution it won't really be distinguishable. 

All the methods have limitaitons:

1) Favorable Viewing Geometry

2) Favorable Viewing Geometry, Large Target, and Deep Target

3) High Resolution or Large Target

4) Large Target and Deep Target

Only the specular reflection doesn't require a large target, and it's favorable viewing geometry limitation is extremely restrictive. We honestly need a new method... which is where *I* come in!
