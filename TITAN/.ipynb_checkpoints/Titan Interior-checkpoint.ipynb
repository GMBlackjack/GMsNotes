Titan's interior can only be examined through indirect methods. This is the case with virtually all planets and moons on the solar system to begin with, but Titan's thick atmosphere makes what little can usually be determined about a body's interior even harder to discern.

# Bulk Properties

The simplest thing we can learn about Titan's iterior is the composition. When we measure Titan we primarily measure its inteiror, so we can find rough estimates of overall composition just from orbital and visual observaitons (which gives us the density; the actual parameter we examine). 



*Mass ratio of rock to rock+ice for outer solar system. [1]*

The above graph shows a plot which compares the rock to rock+ice (silicate mass fraction) of bodies in the outer solar system. The unlabeled x axis is distance from the sun. Titan sits in the middle at roughly 0.55, which means half ice and half rock. Keep in mind that rock includes silicates and metals, while ice includes water ice, but also other ices such as ammonia hydrate, dry ice, and trapped gasses like methane or noble gasses [1]. So the ice/rock ratio is not as immediately helpful as we might have thought.









Basic models indicate that Titan has a rocky core that extends to about 60-65% of its radius, at which point it turns to an icy mantle. It is seen as highly probable that the water in this mantle is in liquid form, making a subsurface ocean on which an icy crust "floats." This subsurface ocean is required if cryovolcanism is active on Titan, for it would be the source of the watery "lava" that spews out of the cryovolcanoes. 

This subsurface ocean could also explain where all the methane in the atmosphere comes from; it could be continually restored by eruptions from the interior ocean, where it is mixed with the water. The water may also be mixed with ammonia, which would lower its melting temperature. Typical models suggest the ocean is a few hundred km thick, with the ice shell 60-100 km thick resting on top of that. 

As you may note from the short size of this section, we don't really know much, especially compared to the atmosphere.
