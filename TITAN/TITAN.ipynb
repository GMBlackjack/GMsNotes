## History

Titan was discovered in 1655 by Christiaan Huygens. For most of the history since then Titan has been observed only from the ground, and very little was known about it, besides the fact that it was a large major moon. 

The next major shift came in 1944 when Gerard Kuiper detected methane absorption bands, indicating it had a substantial atmosphere. Granted, methane turned out to not be very prevalent compared to the actual major component, nitrogen, but the atmosphere was still proven.

Major advancements didn't come until the 1980 Voyager 1 flyby, followed by the 1981 Voyager 2 flyby. This determined that the atmosphere was largely nitrogen, and that Titan was a smoggy orange ball in space that was hard to see through. However, it would be possible to spy through the atmosphere at specific infrared wavelengths, which is precisely what the next mission would do.

Cassini-Huygens entered Saturn orbit on 2004 and soon separated into two components. Cassini continued observing the entire Saturnian system, including Titan in both infrared and radar. Huygens, meanwhile, landed on Titan, giving us great data on the atmospheric profile and surface conditions. Cassini continued sending back data and mapping Titan throughout its mission, uncovering many secrets. 

What secrets, exactly? Well, for a quick overview...

## External Interactions

Titan is by far the largest and most massive moon in the Saturnian system, easily dominating all other objects save for Saturn itself. It's slightly larger than Mercury, though significantly less dense. It's considered a regular satellite, one that likely formed with its parent body (Saturn). 

Titan is a large enough body that it influences and is influenced by things around it. There are many odds and ends that can fit here, but the two notable factoids are:

*Saturn's Rings*: Saturn's rings do not vary much in the vertical direction, but what little variation there is is driven by the Sun's gravity... and Titan's gravity. Titan also induces "spiral bending waves" in the rings, which leads to some of the banded structure. 

*Magnetosphere*: Titan usually sits within Saturn's magnetosphere, but it is only within it barely. When the solar wind acts up, Saturn's magentosphere contracts, and Titan is exposed to the solar wind. When this happens, the field lines drape over Titan and form a bow shock, creating an induced magnetosphere that tails off away from the sun. Numerous ions are created while this occurs, and due to Titan's motion around Saturn the magnetosphere is highly asymmetric as produced ions have a preference for one hemisphere over the other. 

This is, admittedly, the "other" category for things that aren't atmosphere, surface, or interior.

## Atmosphere

Of the three major divisions of Titan: the atmosphere, the surface, and the interior, we know the most about the atmosphere. This is perhaps unsurprising as it's the only part we can see from space and also the part that is blocking most of our other measurements. The blocking comes from two sources: methane, and the SMOG. The methane part we understand pretty well, the smog we do not--don't even know what exactly it's made out of, though we do have some guesses.

From space, a good number of deductions can be made, because the complicated nature of the atmosphere produces a lot of details in the spectrum, and from this spectrum we can make a lot of deductions.



*Thermal infrared spectrum of Titan. [2]*

The presence of methane, CH4, is felt all over the infrared bands here in the infrared spectrum. We can also confirm the presence of the actual major constituent of the atmosphere, N2, simply from observations like this. Naturally a ton of other things are identified as well, but even though some spikes are extreme and others are small, this has nothing to do with the prevalence of the gasses: N2 is largely nonreactive so it doesn't show up much despite being the most common part. That said, it does interact stronger in the ultraviolet range (not pictured). 

Note that most of the "bumps" and "spikes" in the graph are upward pointing. This is called being seen "in emission," as apposed to "in absorption" which would appear downward pointing or troughs. This behavior only occurs when an atmosphere's temperature is increasing with elevation, making the gasses higher up emit more thermal radiation. This gives some clues as to what elevation the material is at: in the case of the larger molecules, this happens to be the stratosphere.

The thickness of the smog makes it somewhat easy to say that the atmosphere *is* dense, though precisely how dense was guesswork until Huygens landed, revealing the surface pressure to be around 1.5 bar--significantly dneser than Earth's 1 bar. Huygens was able to determine a lot more than just air pressure, though, for it obtianed atmospheric samples *and* a temperature profile for the atmosphere. 

As we're currently talking about what's *in* the atmosphere, might as well start there.



*Mass Spectra at different altitudes in Titan's atmosphere with Huygens GCMS [3]*

The tables above show the "counts" Huygens received in its various bins as it sailed through the atmosphere. The top shows readings between 120-130 km altitude, the middle shows a special instrument designed to detect rare gasses, and the bottom panel is the detections at the surface. The plot is logarithmic on the y axis, meaning even though N2 and CH4 look close together on the plot N2 is actually somewhere near a hundred times more prevalent. 

Today, we generally accept that the composition of Titan's atmosphere is like this about 95% N2 (molecular nitrogen), 1.4-4.9% CH4 (methane, which is highly variable depending on local conditions), less than 1% Ne (neon), and 0.4% H2 (molecular hydrogen). There are a bunch of hydrocarbons (molecules made from C and H) at lower densities in the atmosphere. These larger, more complex molecules are generally created when solar photons interact with methane, breaking it up and triggering chemical chain reactions. These heavier molecules are what form the smog around Titan. The formed "gunk" is made out of tholins, but the exact composition of titan tholins is uncertain as the exact photochemistry occuring is unknown. These heavy molecules eventually get heavy enough to fall out of the atmosphere and onto the gorund below, after which we know even less about what chemical reactions may occur. As these heavier molecules contain carbon, they are technically "organic" molecules despite not being produced by life.

Because methane interacts so readily with solar radiation, it is destroyed rather rapidly. The amount of methane in the atmosphere would be destroyed entirley in just 1e7 years, or 10,000,000 years. Which may seem like a long time to us but Titan would have lost its methane hundreds of times over during the lifetime of the Solar System if it wasn't somehow replenishing it. The deduction is that the methane *is* somehow replenished, though the exact mechanism as to how is argued about and dependent on surface processes. 

It is worth noting that the major constituent of Titan's atmosphere, N2, is the exact same as Earth's. Titan, however, has a lack of O2 in the atmosphere, while Earth has around 20%. Earth has methane in its atmosphere too, though not to the same extent as Titan.

So that's what's in the atmosphere: nitrogen, methane, and gunk. But beyond the composition, we also want to know the temperature of the atmosphere at different elevations. Huygens also recorded that as it fell. 



*Atmospheric temperature profile of Titan, [4]*

In the above figure, the dotted line is a theoretical prediction and the jagged line is what Huygens actually saw. The layers are unfortunately not labeled, but the bottom layer is the troposphere, the one above that is the stratosphere, then the mesosphere, and then the thermosphere. Notably the theoretical prediction was pretty good up until the Thermosphere, at which point everything goes haywire. The exosphere is not marked but is present at very high altitudes. The smog forms in the stratosphere, so the haywire craziness at high altitudes doesn't change much about what we *see*. 

Here we have a somewehat clear picture of how Titan's atmosphere changes with altitude. The surface sits at a "warm" 93.7 Kelvin. This is somewhat warmer than would be expected from a simple rocky body, which would have a predicted temperature of 85 K, known as the equilibrium temperature. Perhaps predictably this can be attributed to the greenhouse effect, Titan has a lot of methane and it is an effective greenhouse gas. Taking into account all greenhouse effects, raises the surface temperature by 21 K. But wait, this would imply the surface should be 106 K, and it is not that warm! In truth, there is also an anti-greenhouse effect in play accomplished by the smog in the atmosphere. The smog reflects a lot of light out while also being transparent to some infrared wavelengths, which means the planet can radiate heat away more effecitvely. This effect cools the surface by 9 K overall, which shifts the estimate back down to 97 K. This is notably not exactly what Huygens measured but temperature is espected to vary a little bit across different locations. 

Starting from the surface, as altitude increases the temperature goes down, similarly to how it works on Earth, which is sensible as the atmospheres are very similar in their nitrogen dominance. This is the troposphere. Methane clouds form here, not that you can see them through the smog. 

Then suddenly the temperature starts getting warmer with altitude--which also happens in Earth's stratosphere, though not to the same extreme. The stratosphere is where the smog condenses and is the altitude a human eye would see from Titan orbit. There are actually multiple distinct layers of smog, forming multiple shells of the gunk. 

The stratosphere ends when things start cooling off again, signaling the beginning of the mesosphere. The mesosphere admittedly isn't that interesting in a big-picture sense, but it is the part of the atmosphere where radiation to space becomes effective since the smog isn't in the way. 

The mesosphere arguably ends when the temperature inverts again and starts increasing due to the thin atmosphere being bombarded by the sun's light, signagling the start of the thermosphere. Which as we can see in the graph is not as nicely behaved as the other layers. It is heated considerably by the sun's extreme ultraviolet outbut, but it is also cooled by the presence of HCN raadiative cooling, since the ionic chemistry up here just *happens* to be right to produce lots of HCN. 

Titan does have an exosphere above this, the most distant and most tenuous part of the atmosphere where molecules such as hydrogen sometimes fly off an escape. This place is remarkably stable at 160 K. It's not labeled on the graph but it starts at around 1200 km, though it's so thin it would be notably difficult to get an accurate temperature reading on it. Exospheres are so thin that, even if they were hot, it would feel no different to anyone there than if they were in space itself. 

Along the way, Huygens also measured the wind speed at the various atmospheric layers. The way we did this is rather interesting: Huygens itself had no instrument to measure wind. Instead, we measured Huygens *position* and how fast it was changing *due* to wind speed by examining the Doppler Effect in Huygens' transmissions. From this we found that there is almost no wind at the surface, a few m/s at most, all the way up to 60-80 km altitude, at which point the wind increases to 100m/s speeds.

Huygens did not measure the winds above that level, but using remote sensing we have managed to pin down some speeds. In the upper stratosphere we get 100-200 m/s, while once we get into the mesosphere the speed drops again to about 60 m/s.

Now, let's back up a bit. In the troposphere, beneath the haze, *clouds* will form. These clouds are mostly methane, but ethane clouds have been observed.



*Clouds, https://commons.wikimedia.org/wiki/File:Titan_clouds_by_NIRCam_(2022-11-04_annotated).png,https://commons.wikimedia.org/wiki/File:Titan_clouds_by_NIRCam_and_Keck_NIRC-2_(2022-11-04,06_annotated).png*

Even though the clouds are obscured by the haze layer, we can still peak through the layer using infrared imaging to see the clouds, as in the above fuzzy images. Unlike Earth, which has clouds basically everywhere at all times, the clouds of Titan only cover about 1% of the surface area at any given time, up to 10% during storm events. Otherwise the sky is "clear" (just smoggy). Clouds are more common at the poles than mid-latitudes and equatorial regions, but we have seen clouds appear in every location. The polar prevalence of clouds is thought to be due to both seasonal effects and the presence of the methane lakes, which we will get to when we discuss the surface. 

Huygens data implies the existence of a single "cloud" that surrounds all of Titan's atmosphere around 25-35 km in elevation. However, this "cloud" is so thin that it can't be seen. 

Methane clouds are not the only clouds to form, ethane clouds have been found to form as well, though not as commonly as ethane isn't as large of a part of the atmosphere. They form at about 40km height. 

With clouds, we have finally started to consider the fact that Titan is not actually a perfectly symmetrical ball that is the same everywhere. It experiences day and night tempreature fluctuations, and due to its axial tilt it experiences seasons as well. The highest polar regions have sections that don't see the sun for years, or are always lit *for* years. (Earth years, the Titan year is the same as the Saturnian year, and the seasons change with that). 

This means certain sections of the Titan atmosphere can be in darkness for a long, long time, and receive no radiation from the sun. When this happens, there's no energy to break apart larger molecules and they build up, creating a thick polar cap of smog over whichever pole is currently experiencing winter, meanwhile the summer pole is going to experience far more evaporation and see lots of clouds as well as complex atmospheric circulations. 

This dichotomy between the two poles of Titan means there arises an overall *flow* across Titan. Air rises over the summer pole until it reaches the stratosphere and the smog, at which point it flows all the way across Titan until it descends over the winter pole, only adding to the already intense polar hood. This flow is called meridonial circulation. This overall flow takes the materials in the stratosphere and depoits them into the troposphere. Ethane condenses at the winter pole when this happens, creating ethane clouds, and several other species condense as well. 

Now with all these clouds and weather patterns, there is rain. The smog also gets heavy enough to just fall out of the atmosphere over time. Thus, the study of the atmosphere leads us to the study of the surface of Titan.

## Surface

Due to the smog, for the longest time we didn't know what the surface of Titan was even like, and distant measurements from Earth could only give very vauge pictures. However, once Huygens landed, we got a picture of a single location.



*Contrast-enhanced Huygens image of Titan's surface https://commons.wikimedia.org/wiki/File:Huygens_surface_color_sr.jpg*

Huygens landed on a solid and relatively flat surface. Besides the image Huygens did learn a few other things--mainly, that once it landed, the methan abundance in the air around it increased by 40%, which suggested that there had been liquid methane in the soil that was released upon encountering Huygens. 

Huygens only sampled one location though, the rest of the information came from Cassini as it flew by Titan several times, imaging it with radar and infrared. The smog may block most light and all visible light, but the specific infrared wavelengths that can get through make a striking image. 



*Titan composite surface in IR  https://commons.wikimedia.org/wiki/File:PIA20016-SaturnMoon-Titan-20151113.jpg*

Despite these images, though, as a whole we still don't know what the surface is made out of. It has been noted that the tholin "gunk" that forms in the atmosphere should precipitate down onto the surface over time, pehraps making a layer a few hundred meters thick. But the exact composition of this layer is an unknown, who knows what kinds of chemical processes happen to it while on the surface? 

However, while we do not know *what* the surface is made of, we can still see it and identify particular shapes and land types. In fact, we even have global maps.



*Titan composite surface map, global, with labels. https://commons.wikimedia.org/wiki/File:PIA20713-Titan-SaturnMoon-LabeledFeaturesIAU-June2015.jpg*

The equatorial region of Titan is composed mostly of dune fileds, the dark areas on the above image. In fact, they would go *all* the way around if it wasn't for the highly unusual patch of terrain known as Xanadu, where not a single dune field is to be seen. The weird thing about dunes is that while Titan's winds should usually flow west to east, the dunes are arranged as though it was east to west. 

There aren't many features at the mid-latitudes. 

At the poles things become interesting again, because there are lakes at the north pole and a lake at the south pole, lakes of methane and possibly other constituents. 



*Lakes! The Lakes! Radar image, https://en.wikipedia.org/wiki/File:Liquid_lakes_on_titan.jpg*

Unambiguous lakes have only been found at the polar regions, ranging from massive seas to tiny little lakes. The seas are persistent and longlived, but the small lakes have been observed to change over time. 

In addition to the large features, our scans have found some notable small features as well. Numerous channels clearly cut by liquid have been identified, though there is lots of uncertainty in anything beyond that. Radar-bright channels may be rivers filled with large boulders, while radar-dark ones may be smooth deposits... or filled with liquid, maybe. There's evidence of tributaries and river detlas. However, we have no confirmed low-latitude lakes, so these channels are unlikely to be active.

Titan doesn't have many craters, but it does have a few. The low number of craters indicates that the surface is young and has a low resurfacing rate--estimates place it betweeen 1e8 and 1e9 years old, which is very short on Solar System timescales. 

However, due to viewing limitations it is hard to tell what exactly is a crater and what isn't. Some of the ambiguous possibly-craters could also be cryovolcanoes--that is volcanoes that shoot out water instead of lava. There is no confirmed cryovolanism on Titan, but there are several features that would make sense if they were caused by such activity, including radar-bright apparent "flows." 

Of course, if we ask what drives cryovolcanism, that drives us into Titan's interior...

## Interior

We can actually tell a little bit about Titan's interior from far outside, just measuring its mass and its size. This provides the bulk density, which we can use to determine a very rough composition.



*Mass ratio of rock to rock+ice for outer solar system. [1]*

The above graph shows a plot which compares the rock to rock+ice (silicate mass fraction) of bodies in the outer solar system. The unlabeled x axis is distance from the sun. Titan sits in the middle at roughly 0.55, which means half ice and half rock. Ice is not necessarily water-ice (though it mostly is), it can be other so-called "volatiles" such as ammonia. 

Basic models indicate that Titan has a rocky core that extends to about 60-65% of its radius, at which point it turns to an icy mantle. It is seen as highly probable that the water in this mantle is in liquid form, making a subsurface ocean on which an icy crust "floats." This subsurface ocean is required if cryovolcanism is active on Titan, for it would be the source of the watery "lava" that spews out of the cryovolcanoes. 

This subsurface ocean could also explain where all the methane in the atmosphere comes from; it could be continually restored by eruptions from the interior ocean, where it is mixed with the water. The water may also be mixed with ammonia, which would lower its melting temperature. Typical models suggest the ocean is a few hundred km thick, with the ice shell 60-100 km thick resting on top of that. 

As you may note from the short size of this section, we don't really know much, especially compared to the atmosphere.
