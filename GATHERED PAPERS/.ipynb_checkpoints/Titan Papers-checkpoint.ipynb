## General Properties

**Smith, 1980**: The Radius of Titan From Pioneer Saturn Data

https://agupubs.onlinelibrary.wiley.com/doi/epdf/10.1029/JA085iA11p05943 

Ever wanted to know how to calculate the radius of an object? Well, here you are. Do note that the radius is wrong as at this point we did not know the extent of the atmosphere. It did sucessfully predict the blue "halo" Voyager saw later.

## Atmosphere

**Kuiper, 1944**: Titan: a Satellite With an Atmosphere

https://articles.adsabs.harvard.edu/pdf/1944ApJ...100..378K

The confirmed discovery of Titan's atmosphere using spectroscopy that found methane signatures. Good for historic reference. Also good if you need to cite that Titan is orange, for some reason.

**M. Es-sayeh et al, 2023**: Updated Radiative Transfer Model for Titan in the Near-infrared Wavelength Range: Validation against Huygens Atmospheric and Surface Measurements and Application to the Cassini/VIMS Observations of the Dragonfly Landing Area.

https://iopscience.iop.org/article/10.3847/PSJ/acbd37/meta

This paper is primarily concerned with the description of an updated radiative transfer model for Titan's atmosphere, including new data and some better procedures. It compares the results of its previous model with the new model for the Huygens Landing Site. Of special note is the discovery of the non-integer Fractal Dimension of the aerosols; previous work had used 2, but this work finds that results work best with 2.3 or 2.4 Fractal Dimension. (Quick explanation: fractal dimension is how much an object's "measure" scales with doubling. Squares when doubled go like $x^2$, with a fractal dimension of 2, while cubes go like $x^3$, thus 3. The aerosol surface area is thus structured enough, or "rough" enough, to scale faster than a simple 2D model would imply). There is a lot of discussion on how to remove seams in VIMS images due to viewing geometry differences, and the model is also applied to Selk Crater and surrounding areas, including the Dragonfly Landing Site. This paper is a bit too crammed full of stuff, in my opinion, and could have been split up. Regardless, the ultimate goal of the model was to extract surface albedo and assign labels to different terrain types, which it appears to have succeeded at. 

**Barnes et al, 2018**: Titan's Twilight and Sunset Solar Illumination

https://iopscience.iop.org/article/10.3847/1538-3881/aae519/meta

A great example paper of using SRTC++ in an actual application, especially one that's not the most obvious--using sections of Titan's surface as the detectors. It generates what Titan susnets would look at from different wavelengths, including visible. (Wait, weren't we not very good otuside of the IR windows? Check with Dr. Barnes about this). Also just nice pictures.

**Barnes et al, 2020**: Diffraction-limited Titan Surface Imaging from Orbit Using Near-infrared Atmospheric Windows

https://iopscience.iop.org/article/10.3847/PSJ/ab91b6/meta

Compares a SRTC++ code with an analytic code for how high of a resolution could we expect to get from a Titan orbiter. Turns out, answer is about 20 meter pixels. Shows that SRTC++ has some improvements over the analytical version largely due to the inclusion of the forward-scattering function. Also splits up components of what is detected from direct-from-surface to atmospheric to surface-then-atmospheric-scattering. The differneces are striking at different wavelengths.

**Negrao et al, 2005**: On the Latitudinal Distribution of Titan's Haze at the Voyager Epoch

https://www.sciencedirect.com/science/article/pii/S0032063304002545

Shows how Voyager's data pointed out the variation in Titan's haze from the polar regions to the equatorial. Talks a lot about modeling it all. Honestly not entirely sure this one's worth saving here, but here we are anyway.

**Brown et al, 2002**: Direct Detection of Variable Tropospheric Clouds Near Titan's South Pole

https://www.nature.com/articles/nature01302

So this paper has a few interesitng things. First, direct images of Titan clouds, a first. Second, recognition that the clouds congregate near poles. Third, a hilariously wrong thought is mentioned--that the clouds should be congregating near the equator, perhaps eternally! No, this ain't right. The paper does go to discuss that the previous theories may be wrong, but I do find it amusing. Oh, fourth, a nice example of adaptic optics power.

## Surface (in general)

### y'know, not specifically lake stuff

**Lopes et al, 2020**: A Global Geomorphologic Map of Saturn's Moon Titan

https://www.ncbi.nlm.nih.gov/pmc/articles/PMC7271969/

This is the paper with all the terrain maps splitting Titan up into six types of areas: lakes, craters, dunes, hummocky stuff, labyrinths, and plains. Notably craters are questionable as a "terrain type" and only some of the lakes are actually filled with liquid. Good place to refer to basic descriptions of each of these. Excellent map to boot, though not very high resolution, and can be questionable in interpretaion. A good general place to go. 

**Malaska et al, 2016**: Geomorphological Map of the Afekan Crater Region, Titan: Terrain Relationships in the Equatorial and Mid-latitude Regions.

https://www.sciencedirect.com/science/article/pii/S0019103516000889 

This is a paper concerned with a single crater but that's not why we use it. Sure, Afekan Crater, you're very interesting, but this paper's main use is the fact that it heavily subdivides the major terrain types of Titan into many different types, each with their own unique section that describes them. Contains labyrinths, but not lakes. there aren't lakes near Afekan Crater. This is a more fine-tuned look at topography compared with the above paper.

**Solomonidou et al, 2018**: The Spectral Nature of Titan's Major Geomorphological Units: Constraints on Surface Composition.

https://agupubs.onlinelibrary.wiley.com/doi/full/10.1002/2017JE005477

An alternative look at the types of terrain on Titan, though this one does not attempt to make a map, rather it attempts to characterize and identify the various spectral properties of the topographic units. This allows some tentative material identification, though I mean tentative. We can be pretty sure there is water ice in some of the locations, particuarly at higher latitudes. Beyond that we raelly aren't sure, the equator has organics, but what even are those organics? The paper puts up a few theories as to why this disparity in composition with latitude exists.

**Lopes et al, 2016**: Nature, Distribution, and Origin of Titan's Undifferentiated Plains

https://www.sciencedirect.com/science/article/pii/S0019103515005576

Discussion of the Blandlands. The empty boring planes. You'd think it would be the simple kind of terrain but NO it's VERY CONFUSING and has all SORTS of issues. The conclusion in this paper is that the plains are sedimentary in origin and primarily driven by wind processes, but there are other possible contributing factors. It also relates the plains to other types of topography, such as the dunes and labyrinths. 

**Lorenz et al, 1995**: Prediction of Aeolian Features on Planets: Application to Titan Paleoclimatology

https://agupubs.onlinelibrary.wiley.com/doi/epdf/10.1029/95JE02708

This paper is only important because the conclusion it draws, "dunes on Titan are unlikely," is completely wrong and disproven by the next paper on this list. By the same researcher.

**Lorenz et al, 2006**: The Sand Seas of Titan: Cassini RADAR Observations of Longitudinal Dunes

https://www.science.org/doi/10.1126/science.1123257

Discovery paper for the Dunes! Discovered via Cassini Radar. Talks about how to identify dunes from radar images. Naturally there are better papers to characterize the dunes, this is just the discovery one.

**Malaska et al, 2020**: Labyrinth Terrain on Titan

https://www.sciencedirect.com/science/article/pii/S0019103520301524

Paper that discusses the Labyrinths. Goes into detail, subdivides the labyrinth, draws attention to the unusual Kronin Labyrinthus region. Radar-focused paper with lots of good images and explanations, as well as a global (at the time) mark of all known labyrinths. Goes into a truly stupendous amount of detail, surely it's useful somehow, right? So much labyrinth information. 

**Soderblom et al, 2010**: Geology of the Selk Crater Region on Titan from Cassini VIMS Observations

https://www.sciencedirect.com/science/article/pii/S0019103510001028

Selk Crater is interesting and worthy of being looked at. That's what this paper does, largely topographically, examining both the crater itself and the Bench. This crater is notably extra important since it's the one Dragonfly is visiting but this paper isn't concerned with that. We don't really know what the Bench is but two theories are discussed: fluidized ejecta flow or a streamlined upland flow due to erosion. Notably the Bench looks to potentially be made *overtop* an ancient crater highland. Lots of mysteries afoot.

## Lakes

**Stofan et al, 2007**: The Lakes of Titan

https://www.nature.com/articles/nature05438

This paper is very simple but very important. It's the DISCOVERY of the lakes on Titan. Woot! Uses radar images. 

**Hayes, 2016**: The Lakes and Seas of Titan

https://www.annualreviews.org/doi/pdf/10.1146/annurev-earth-060115-012247

An excellent summary of research concerning the lakes and seas of Titan. As a review it probably shouldn't be cited itself, but it can point to other citations, and goes into enough detail to showcase what facts about the seas we know and which ones are particuarly relevant / the subject of current study. 

**Vixie et al, 2015**: Possible Temperate Lakes on Titan.

https://www.sciencedirect.com/science/article/pii/S0019103515002110

The only confirmed lakes of Titan are on the poles, but here we have evidence of *possible* lakes in mid-latitude regions. The actual identity is still shrouded in mystery, and this paper is part of the reason we need better radiative transfer models so we can determine if these are really lakes or not. Has locations of many potential candidates, and includes a discusison on Griffith et all 2012's equatorial candidate. This paper can't rule it out entirely but it does not think the equatorial candidate has the right properties.

**Griffith et al, 2012**: Possible Tropical Lakes on Titan From Observations of Dark Terrain.

https://www.nature.com/articles/nature11165

A mysterious dark spot on Titan that might be an equatorial lake. However, other papers I read cast doubt on this claim, but it is a possible candidate for later research once we have a radiative transfer model running. 

## History

**Lorenz and Mitton, 2002**: Lifting Titan's Veil: Exploring the Giant Moon of Saturn

https://web.archive.org/web/20050222073209/https://assets.cambridge.org/052179/3483/sample/0521793483ws.pdf

Excellent source of basic history of Titan's discovery and various advances up until before the Cassini mission. This is just the first chapter of a larger book and yet I have not read the larger book--this gives an excellent overview though, sufficient to understand the path Titan has taken through history.
