### 1.3 microns

1.3 microns shows that the HLS has pretty close to 0.1 albedo. The HLS data matches the simulation extremely closesly across all places the data exists: the only discernable deviation is a potential inverted slope when slicing across emission: low emission values are brighter than higher emission values for the real data, meanwhile the simulation is increasing over that range. This could simply be a case of seeing patterns where there's not enough data to say for sure.

## 2 microns

Similarly to 1.3 microns, the HLS data matches up well with the model at 0.1 albedo. However, we still see in many emission slice views the increase in HLS intensity (I/F) at low emission. Notably, we do not always see it: our example emission slice shows a decrase in intensity at low emission, adding credence to the idea that this uptick is just noise.

What isn't just noise is the clear preference on the azimuth-slices for large azimuth angles to have more intense I/F. Note that while this happens at 2 microns, it does not at 1.3 microns! A discovery? (This uptick does not exist in all slices; there may be a preference for lower emission/incidence angles)

EXTRA: just to show that the emission slice usually still ticks up, not down, at low emission.



## 5 microns

To put it simply, we expect 5 microns to be bizarre as it doesn't follow the same patterns the other windows do.

As the 5 micron window doesn't lie on 0.1 albedo like the others, this is makes sense. The shape seems more-or-less in line with what the model expected, though since the shape of the 5 um window's model's change so much this is not as clear cut as other windows, with the exception of the azimuth slice. Whether this implies that the HLS is just particuarly dark at 5um or if the model is not capturing the intricacies of the 5um window is unclear.

The higher-azimuth higher-intensity trend shown in the 2um window also occurs here. The "uptick" at low emission in emission slices is not present in most views.
