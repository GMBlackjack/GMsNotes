```python
# --- THE ONE AND ONLY STARTUP CELL ---
# Run this in a fresh kernel before anything else.

import os

# 1. Fix the environment BEFORE Matplotlib is ever imported.
# This removes the "Original Sin".
if 'MPLBACKEND' in os.environ:
    del os.environ['MPLBACKEND']

# 2. Now, with a clean environment, activate the interactive backend.
# This will be the FIRST time this kernel imports Matplotlib.
# Because the environment is clean, it will import correctly and not "poison the well".
%matplotlib ipympl

import matplotlib.pyplot as plt
import numpy as np

print("Environment fixed and ipympl backend activated successfully.")
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)
ax.plot(x, y)
ax.set_title("Interactive Test Plot")
plt.show()
```
