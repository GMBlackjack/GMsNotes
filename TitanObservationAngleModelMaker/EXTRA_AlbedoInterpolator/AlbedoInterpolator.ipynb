```python
ani.pause()
plt.close()

plt.rcParams['animation.ffmpeg_path'] = r"C:\Users\deran\ffmpeg-2025-12-10-git-4f947880bd-essentials_build\ffmpeg-2025-12-10-git-4f947880bd-essentials_build\bin\ffmpeg.exe"

#2D PLOT!
#The idea is to animate through an entire model, starting at incidence 0 and going to incidence 100, with emission and azimuth forming a CIRCULAR 2D plot. Great for visualization!
#FEATURES:
#Basic animating 2D Plot
#Markers for Real Points

%matplotlib ipympl
# Set the model by setting this line here to the one you want to look at:

import copy

modelPick = 7
animMatrix = loadedModels[modelPick+24]
animCounts = loadedModelCounts[modelPick+24]
animAlbedo0 = loadedModels[modelPick]
animAlbedo1 = loadedModels[modelPick+8]
animAlbedo2 = loadedModels[modelPick+16]
#insert the number of the model you wish to display.

mcmap = copy.copy(plt.get_cmap('gist_ncar'))
#mcmap = copy.copy(plt.get_cmap('bwr'))
mcmap.set_under(color = 'black')
#if you want to change the color map

vmin=-0.1
#Recommended to set this to 0 and never touch it. 

vmax=0.3 #Maximum displayed value.
#Raise if you're saturating, lower if you can't see any detail.

Z = 21 #incidence
Y = 19 #Emission
X = 72 #Azimuth because polar coordinates think angle is x in matplotlib for some reason. 

fig, ax = plt.subplots(figsize = (12,12),subplot_kw=dict(projection='polar'))
ax.set_theta_zero_location("N") #"0" degrees is DOWN. 

#Now we'd like to try to animate it. 
#https://matplotlib.org/stable/users/explain/animations/animations.html
#Eventually had to use ChatGPT to help here. Could not get the ArtistAnimator to work, but it could
#get the FuncAnimation to work--evidentially defining zz inside the funciton/loop was important.
#Even though I thought I already tried that...

#Now we actually need to get the slices
#Set up x and y divisions.        
yy = np.linspace(0, 90, 19)
xx = np.linspace(0, np.deg2rad(355), 72) #We will mirror the points across to fill the full view.

XX, YY = np.meshgrid(xx,yy)

c = ax.pcolormesh(xx,yy, np.zeros((Y, X)), cmap=mcmap,vmin=vmin,vmax=vmax)
fig.colorbar(c, ax=ax, label='Retrieved Albedo')
# annoyance of animation: trying to update the color bar duplicates it.
# so we need a "dummy" plot to initialize it before we start animating. 

#Making an array so the title can display the azimuth angle. 
angles = ["000", "005", "010", "015", "020", "025", "030", "035", "040", "045", "050", "055", "060", "065", "070", "075",
             "080", "085", "090", "095", "100"] 

def tripleAlbedoRetrieve(measurement,albedo0,albedo1,albedo2):
    #When calling this function, we assume...
    #First, the simple ones are just linear interpolaton. 
    foundAlbedo = 1
    if ((measurement > albedo0) and (measurement < albedo1)):
        foundAlbedo = 0.0 + 0.1*((measurement-albedo0)/(albedo1-albedo0)) #0.0 to 0.1
    elif ((measurement > albedo1) and (measurement < albedo2)):
        foundAlbedo = 0.1 + 0.1*((measurement-albedo1)/(albedo2-albedo1)) #0.1 to 02
    elif (measurement == albedo0):
        foundAlbedo = 0.0
    elif (measurement == albedo1):
        foundAlbedo = 0.1
    elif (measurement == albedo2):
        foundAlbedo = 0.2 #I highly doubt data will ever be EXACT like this, but we need to catch the possibility. 
    else: #Now we have to EXTRAPOLATE. Use all three points to get the most data. 
        coefficients = np.polyfit([albedo0, albedo1, albedo2], [0.0,0.1,0.2], 2) #last number is the Order here. 1 is linear.
        equation = np.poly1d(coefficients)
        foundAlbedo = equation(measurement)
    return foundAlbedo

def update(frame):
    ax.clear()
    zz = np.zeros((Y, X))
    cc = np.zeros((Y, X))
    aa0 = np.zeros((Y, X)) #hmm these probalby didn't need to be matrices actually come to think of it... whoops.
    aa1 = np.zeros((Y, X))
    aa2 = np.zeros((Y, X))
    for x in range(X):
        for y in range(Y):
            if (x <= 36): #Check exists to "mirror" data around the circle. 
                zz[y][x] = animMatrix[frame][y][x]
                cc[y][x] = animCounts[frame][y][x]
                aa0[y][x] = animAlbedo0[frame][y][x]
                aa1[y][x] = animAlbedo1[frame][y][x]
                aa2[y][x] = animAlbedo2[frame][y][x]
            else:
                zz[y][x] = animMatrix[frame][y][36-(x-36)]
                cc[y][x] = animCounts[frame][y][36-(x-36)]
                aa0[y][x] = animAlbedo0[frame][y][36-(x-36)]
                aa1[y][x] = animAlbedo1[frame][y][36-(x-36)]
                aa2[y][x] = animAlbedo2[frame][y][36-(x-36)]
            #Now we need to retrieve an albedo from this. We cannot assumed the albedos are going to behave physically, though, as the simluations are monte carlo.
            #First off if we simply have no data... just nuke it.
            if (zz[y][x] > 0):
                if (aa0[y][x] < aa1[y][x] and aa1[y][x] < aa2[y][x]): #Points are physical, be normal!
                    zz[y][x] = tripleAlbedoRetrieve(zz[y][x],aa0[y][x],aa1[y][x],aa2[y][x])
                else: #Something freaky has happened.
                    if (aa1[y][x] == 0 and aa0[y][x] < aa2[y][x]): #missing data point is fine, go ahead and interpolate with just the two. This also works as extrapolation if necessary!
                        zz[y][x] = 0.0 + 0.2*((zz[y][x]-aa0[y][x]) / (aa2[y][x]-aa0[y][x]))
                    elif (aa2[y][x] == 0 and aa0[y][x] < aa1[y][x]): #missing data point is fine, go ahead and interpolate with just the two.
                        zz[y][x] = 0.0 + 0.1*((zz[y][x]-aa0[y][x]) / (aa1[y][x]-aa0[y][x]))
                    #Notably, if aa0==0, then the normal albedo retrieve function will work just fine and we won't even be here.
                    elif (aa0[y][x] == 0 and aa1[y][x] == 0 and aa2[y][x] == 0):
                        zz[y][x] = np.nan #There's no data here.
                    else: #Panic, there's data here but it doesn't make any sense. Whoopsy kinda thing. 
                        zz[y][x] = np.nan
            else:
                zz[y][x] = np.nan
            if (zz[y][x] < vmin): zz[y][x] = vmin
            if (np.isnan(zz[y][x])): zz[y][x] = -10
    azimuth = ax.pcolor(xx, yy, zz, cmap=mcmap,vmin=vmin,vmax=vmax)
    mask = cc>0
    ax.scatter(XX[mask],YY[mask],s=15,c='white', edgecolors='black', linewidth=0.5, alpha=0.8, label='Real Data')
    ax.set(xlim=[0,6.28], ylim=[2.5,90], 
           title = '5.00um Equatorial Dunes Restricted BDRF. Incidence: '+str(angles[frame])+ ' degrees')
    ax.set_theta_zero_location("N") #"0" degrees is DOWN. 
    ax.tick_params(axis='y', colors='red') # The radial numbers (0, 20, 40...)
    ax.grid(color='gray',linewidth=1)
    for spine in ax.spines.values():
        spine.set_edgecolor('none')

ani = animation.FuncAnimation(fig, update, frames=Z, interval=200)

# To save the animation using Pillow as a gif. Use FFMpegFileWriter for mp4. 
writer = animation.FFMpegFileWriter(fps = 0.5, bitrate=1800)
ani.save('RestrictedBDRFAlbedos5.00.mp4', writer = writer, dpi=200)

plt.show()
```

```python
#Time for a more general albedo retriever. We create a full "restricted BDRF" of every wavelength, except that it has the retrieved albedo values. 
#Then we find the AVERAGE retrieved albedo for each incidence, emission, and azimuth bin. This should be fun. 

fullAlbedos = [[np.zeros((Y, X)) for z in range(Z)] for w in range(8)] #so the indices will go [w][z][y][x] or [wavelength][incidence][emission][azimutn]

smoothenOutliers = True #Outliers really screw up final printing sometimes, use this to chill them out a tad.

aa0 = 0
aa1 = 0
aa2 = 0
for w in range(8):
    for z in range(Z):
        for x in range(X):
            for y in range(Y):
                if (x <= 36): #Check exists to "mirror" data around the circle. 
                    fullAlbedos[w][z][y][x] = loadedModels[w+24][z][y][x]
                    aa0 = loadedModels[w][z][y][x]
                    aa1 = loadedModels[w+8][z][y][x]
                    aa2 = loadedModels[w+16][z][y][x]
                else:
                    fullAlbedos[w][z][y][x] = loadedModels[w+24][z][y][36-(x-36)]
                    aa0 = loadedModels[w][z][y][36-(x-36)]
                    aa1 = loadedModels[w+8][z][y][36-(x-36)]
                    aa2 = loadedModels[w+16][z][y][36-(x-36)]
                if (fullAlbedos[w][z][y][x] > 0):
                    if (aa0 < aa1 and aa1 < aa2): #Points are physical, be normal!
                        fullAlbedos[w][z][y][x] = tripleAlbedoRetrieve(fullAlbedos[w][z][y][x],aa0,aa1,aa2)
                    else: #Something freaky has happened.
                        if (aa1 == 0 and aa0 < aa2): #missing data point is fine, go ahead and interpolate with just the two. This also works as extrapolation if necessary!
                            fullAlbedos[w][z][y][x] = 0.0 + 0.2*((fullAlbedos[w][z][y][x]-aa0) / (aa2-aa0))
                        elif (aa2 == 0 and aa0 < aa1): #missing data point is fine, go ahead and interpolate with just the two.
                            fullAlbedos[w][z][y][x] = 0.0 + 0.1*((fullAlbedos[w][z][y][x]-aa0) / (aa1-aa0))
                        #Notably, if aa0==0, then the normal albedo retrieve function will work just fine and we won't even be here.
                        elif (aa0 == 0 and aa1 == 0 and aa2 == 0):
                            fullAlbedos[w][z][y][x] = np.nan #There's no data here.
                        else: #Panic, there's data here but it doesn't make any sense. Whoopsy kinda thing. 
                            fullAlbedos[w][z][y][x] = np.nan
                else:
                    fullAlbedos[w][z][y][x] = np.nan
                if (smoothenOutliers == True):
                    if (fullAlbedos[w][z][y][x] < -1): fullAlbedos[w][z][y][x] = -1
                    if (fullAlbedos[w][z][y][x] > 1): fullAlbedos[w][z][y][x] = 1

#REMEMBER FOR THE AVERAGES: AZIMUTH IS COUNTED TWICE, MAKE A DECISION ON HOW TO DEAL WITH IT!
#print(fullAlbedos[4])

inciTrends = [[0 for z in range(Z)] for w in range(8)]
emisTrends = [[0 for y in range(Y)] for w in range(8)]
azimTrends = [[0 for x in range(37)] for w in range(8)] #We are NOT doubling up. 

for w in range(8):
    #Get the incidence trend
    for i in range(Z):
        total = 0.0
        counts = 0.0
        for x in range(37):
            for y in range(Y):
                if (not np.isnan(fullAlbedos[w][i][y][x])):
                    total = total + fullAlbedos[w][i][y][x]
                    counts = counts + 1
        inciTrends[w][i] = total / counts
    #Get the emission trend
    for i in range(Y):
        total = 0.0
        counts = 0.0
        for x in range(37):
            for z in range(Z):
                if (not np.isnan(fullAlbedos[w][z][i][x])):
                    total = total + fullAlbedos[w][z][i][x]
                    counts = counts + 1
        emisTrends[w][i] = total / counts
    #Get the azimuth trend
    for i in range(37):
        total = 0.0
        counts = 0.0
        for y in range(Y):
            for z in range(Z):
                if (not np.isnan(fullAlbedos[w][z][y][i])):
                    total = total + fullAlbedos[w][z][y][i]
                    counts = counts + 1
        azimTrends[w][i] = total / counts

print(inciTrends)
print(emisTrends)
print(azimTrends)
```
