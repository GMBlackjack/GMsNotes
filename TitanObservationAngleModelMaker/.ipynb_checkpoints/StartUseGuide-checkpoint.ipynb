# Titan Observation Angle Model Maker

## Start Use Guide

### Make models, display models, compare models, this folder can do it all!

Hello and welcome! This is the Start Use Guide for the Titan Observation Angle Model Maker. Which I need to give a proper name too because TOAMM looks silly, but whatever.

This folder contains jupyter notebooks and python code that can search the (paired down) VIMS Cube Database in order to compile every single instance a particular location or set of locations on Titan was viewed. The scope of the search can be as narrow as the Huygens Landing Site, or as borad as every single location on Titan. (Though why you would want that last one, I have no idea.) We even have a built-in mask that can be used to differentiate Titan's major topographical units; such as the Dunes or the Lakes. After the database search is complete, we also have programs that can compile the recorded observations and sort them by observation angle, interpolating between points to create a model of the terrain across many viewing angles. Once a model is created, it can be compared to other models in yet another notebook, and this is where most of the science is done.

Code to harvest similar data from simulated Titans from SRTC++ is also avialable, so they can be compared with the models that have real data.

If you just want to compare models that already exist, then direct yourself to the **ModelViewer** notebook, it has all the informaiton you need.

Creating your own models is a bit more complicated. If you want to start from scratch, then go to the **DatabaseSearcher**. It will, from a set of given ctieria, create a list of cubes that satisfy said criteria. 

If you already have a list of cubes, head to **PixelFinder**. This will take the cubes in a provided list and find specific pixels that match a set of criteria, and then it will print out said pixels (and a lot of associated data with them) to a new file. This is intended to be run multiple times with different criteria on the same cube list, rather than searching the entire database every time.

Once you have a pixel list you turn to **ModelCreator** which will take the pixel list and create a model out of it. There are options to sift the pixels further at this point, if you want; maybe the PixelFinder didn't differentiate between types of terrain but you want to now. Up to you.

Then of course once the model is created return to **ModelViewer** to have a look at it and compare it to other models. 

There are also multiple kinds of files in this folder that are not notebooks. These are as follows...

AcceptableCubes.csv: A file that stores the names of all acceptable cubes in the database so it doesn't have to be fetched every time. 

...CubeList.csv: A list of cubes generated by DatabaseSearcher (or perhaps some other means). Only contains flyby and cube number information.

CLRMaskArray.npy and CLRMaskArrayDist.npy are numpy files that hold arrays that serve as masks of titan to help differentiate terrain types from each other, and also how certain we are something saying it's in that pixel really is that terrain.

...PixelResults.csv: A list of pixels generated by PixelFinder. Contains a lot of information about every pixel in the list; files can be large.

....npy: .npy files hold the models themselves, they are the results. Their titles are generally of the form NAME#.##umTERRAIN.npy, where NAME is just whatever label was applied to it, the #.##um specifices the wavelength, and TERRAIN the... terrain. the TERRAIN is sometimes left off if it is not specified when the model is constructed.
