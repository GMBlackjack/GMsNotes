The Sleeping Beauty experiment is like this:

A coin is flipped. It is flipped only once, in secret. 

The test subject wakes up one day and asked whether the coin came up heads or tails. 

If the coin came up tails, the test subject is put to sleep and is injected with a memory wiping agent. Then the test subject is asked if the coin comes up heads or tails again.

This is the default version of this experiment. Here, if the coin comes up heads the test subject is asked the question once, and if the coin comes up tails the test subject is asked the question twice. 

We seek to answer two questions here:

1) what is the best strategy for guessing correctly the largest number of times?

2) what is the best stretegy for guessing correctly the largest proporiton of times? That is, per experiment basis, not per time the question is asked.

There is actually a third more philosophical question.

3) if you are the test subject, what should you believe has happened? Tails or Heads? 

The code answers 1 and 2, 3 is a bit more... fun. 

```python
import math
import random 

def flipCoin(x):

    flip = random.random()
    coin = 0
    if (flip < x): 
        coin = 0
    elif (flip > x):
        coin = 1
    else:
        print("How did you do that...?")
        # seriously, landing directly on the probability?? What??

    return coin

# How many tests to run per strategy. 

tests = 1000000

# First strategy, always guess heads. 

i = 0
correctGuesses = 0.0
totalGuesses = 0.0
correctExperiments = 0.0
while (i < tests):
    flip = flipCoin(0.5)
    if (flip == 0):
        correctGuesses = correctGuesses + 1.0
        totalGuesses = totalGuesses + 1.0
        correctExperiments = correctExperiments + 1.0
    else:
        totalGuesses = totalGuesses + 2.0
    i = i + 1

print("Always Guessing Heads")
print(correctGuesses/totalGuesses) # This is the fraction of correct guesses, or the answer to the "betting" probability. 
print(correctExperiments/1000000.0) # This is the entire set of experiments, or the probability/fraction of correctness per experiment.

#always guess tails

i = 0
correctGuesses = 0.0
totalGuesses = 0.0
correctExperiments = 0.0
while (i < tests):
    flip = flipCoin(0.5)
    if (flip == 0):
        totalGuesses = totalGuesses + 1.0
    else:
        correctGuesses = correctGuesses + 2.0
        totalGuesses = totalGuesses + 2.0
        correctExperiments = correctExperiments + 1.0
    i = i + 1

print("Always Guessing Tails")
print(correctGuesses/totalGuesses)
print(correctExperiments/1000000.0)

i = 0
correctGuesses = 0.0
totalGuesses = 0.0
correctExperiments = 0.0
while (i < tests):
    flip = flipCoin(0.5)
    if (flip == 0):
        flipGuess = flipCoin(0.5)
        if (flipGuess == flip):
            correctGuesses = correctGuesses + 1.0 
            totalGuesses = totalGuesses + 1.0
            correctExperiments = correctExperiments + 1.0
        else:
            totalGuesses = totalGuesses + 1.0
    else:
        j = 2
        tempGuess = 0.0
        while (j > 0):
            flipGuess = flipCoin(0.5)
            if (flipGuess == flip):
                correctGuesses = correctGuesses + 1.0
                tempGuess = tempGuess + 1.0
                totalGuesses = totalGuesses + 1.0
            else:
                totalGuesses = totalGuesses + 1.0            
            j = j - 1
        correctExperiments = correctExperiments + tempGuess/2.0
    i = i + 1

print("Truly Random Response")
print(correctGuesses/totalGuesses)
print(correctExperiments/1000000.0)

# And now the adjusting probability scales.

k = 0.05
sneakyNumber = 2.0 # this is just the nubmer of times the "tails" loop iterates.
while (k < 1.0):
    i = 0
    correctGuesses = 0.0
    totalGuesses = 0.0
    correctExperiments = 0.0
    while (i < tests):
        flip = flipCoin(0.5)
        if (flip == 0):
            flipGuess = flipCoin(k)
            if (flipGuess == flip):
                correctGuesses = correctGuesses + 1.0 
                totalGuesses = totalGuesses + 1.0
                correctExperiments = correctExperiments + 1.0
            else:
                totalGuesses = totalGuesses + 1.0
        else:
            j = sneakyNumber
            tempGuess = 0.0
            while (j > 0):
                flipGuess = flipCoin(k)
                if (flipGuess == flip):
                    correctGuesses = correctGuesses + 1.0
                    tempGuess = tempGuess + 1.0
                    totalGuesses = totalGuesses + 1.0
                else:
                    totalGuesses = totalGuesses + 1.0            
                j = j - 1
            correctExperiments = correctExperiments + tempGuess/sneakyNumber
        i = i + 1

    print("Response with weight ", k)
    print(correctGuesses/totalGuesses)
    print(correctExperiments/1000000.0)   
    k=k+0.05
```

So for case 1) you should always say tails. For case 2) it doesn't matter at all.

What of case 3 though? If you wake up as a test subject and are asked the question, what path should you assume you are on? Your chances of guessing right at any given time on waking up are always 50/50, that hasn't changed. But in all possibilities, you are asked the quesiton once on heads, and twice on tails. The thing is do you consider the time you spend to be a variable? In that case, you choose tails always since any given wake up you are more likely to be in a tails than a heads, especially if you do the experiment multiple times.
