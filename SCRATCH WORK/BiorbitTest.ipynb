```python
import math
import numpy as np
import matplotlib.pyplot as plt 

def funnyfunc(inputs):
    r1 = 2.38e8 #meters
    r2 = 2.12e8 #meters
    rb = inputs
    a1 = (r1+rb)/2.
    a2 = (r2+rb)/2.
    GMs = 3.79e16 #meters^3/seconds^2
    Dv1 = np.sqrt(((2*GMs)/r1) - (GMs/a1)) - np.sqrt(GMs/r1)
    Dv2 = np.sqrt(((2*GMs)/rb) - (GMs/a2)) - np.sqrt( ((2*GMs)/rb) - (GMs/a1))
    return (Dv1+Dv2)

def funnyfunc2(inputs):
    r1 = 2.38e8 #meters
    r2 = 2.12e8 #meters
    rb = inputs
    a1 = (r1+rb)/2.
    a2 = (r2+rb)/2.
    GMs = 3.79e16 #meters^3/seconds^2
    Dv1 = np.sqrt(((2*GMs)/r1) - (GMs/a1)) - np.sqrt(GMs/r1)
    Dv2 = np.sqrt(((2*GMs)/rb) - (GMs/a2)) - np.sqrt( ((2*GMs)/rb) - (GMs/a1))
    return (np.abs(Dv1)+np.abs(Dv2))

print(funnyfunc(400000000))

# 2. Generate data for plotting
# Create an array of 400 x-values from -10 to 10
x_values = np.linspace(2e8, 3e8, 40000)
#x_values = np.linspace(0, 1e9, 40000)
# Calculate the corresponding y-values
y_values = funnyfunc(x_values)
y_values2 = funnyfunc2(x_values)

# 3. Create the plot
plt.figure(figsize=(8, 6)) # Create a figure with a specific size
plt.plot(x_values, y_values) # Plot the data
plt.plot(x_values, y_values2) # Plot the data

# 4. Customize and show the plot
plt.title("Delta-v") # Add a title
plt.xlabel("secondary radius") # Add x-axis label
plt.ylabel("Delta-v") # Add y-axis label
plt.grid(True) # Add a grid
plt.legend() # Display the legend
plt.show() # Display the plot
```

```python
import matplotlib.pyplot as plt

def plot_ordered_bar_chart(data_array):
    """
    Generates and displays a bar chart from an ordered list of numbers.

    Each number in the list becomes a bar, plotted in the order it appears.

    Args:
        data_array (list): A list of numbers (integers or floats).
    """
    # Check if the input list is empty
    if not data_array:
        print("Input array is empty. Nothing to plot.")
        return

    # Create x-axis positions, which are simply the indices of the array
    # e.g., for an array of length 5, this will be [0, 1, 2, 3, 4]
    x_positions = range(len(data_array))

    # Create a figure and a set of subplots
    fig, ax = plt.subplots()

    # Create the bar chart
    ax.bar(x_positions, data_array, color='skyblue')

    # Add labels and a title for clarity
    ax.set_xlabel("Increasing Mass ---->",fontsize=30)
    ax.set_ylabel("Concentration",fontsize=30)
    ax.set_title("")
    
    # Set the x-axis ticks to correspond to the original indices
    # This ensures that for a long list, not every single index is shown,
    # keeping the x-axis readable.
    ax.set_xticks(x_positions)
    ax.set_yticks([])
    ax.set_xticklabels("")
    ax.set_yticklabels("")


    # Display the plot
    plt.tight_layout() # Adjusts plot to prevent labels from overlapping
    plt.show()

# --- Example Usage ---
if __name__ == "__main__":
    # You can replace this with any list of numbers
    sample_data = [5, 12, 8, 18, 7, 14, 11]
    
    # Call the function with the sample data
    plot_ordered_bar_chart(sample_data)

    # Another example with a different number of items
    another_sample = [45, 23, 67, 51, 88, 32, 70, 59, 42, 91]
    plot_ordered_bar_chart(another_sample)
```
