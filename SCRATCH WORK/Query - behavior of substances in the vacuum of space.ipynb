## Why doesn't Hydrogen Solidify in Space?

So obviously in the early universe space wasn't as cold so hydrogen couldn't solidify. But in the modern universe, space is at 3 K. Surely some hydrogen should cool to this temperature and solidify? One would think in rather large quantities, too.

In fact, there are places in the universe even colder than the background of space. The Boomerang Nebula *absorbs* light from the CMB [1]. So perhaps asking *why* it's so cold will give us some hint as to what's going on? The nebula has to be primarily hydrogen, it's a planetary nebula coming from a star shedding its outer layers. (Which should be pretty hot come to think of it, why's it so cold?) 

(Okay so the Boomerang Nebula is apparently classified as a Proto Planetary Nebula, a.k.a. it hasn't fully developed yet). 

Anyway, the temperature measure appears to be occuring because the gas itself is expanding dramatically, which lowers its temperature. PV=nRT after all. Unfortunately for an expanding gas all of these things are changing... we can certianly analyze them though. Volume is going up, Pressure is going down, that we know. How do these things scale with radius?

Volume is easy, it scales with $r^3$. 

Pressure is a little different. It's force over area. In a noble gas, we can imagine the particles moving outward and not touching each other, and an invisible wall at the edge. The surface area of the sphere goes up proportional to $r^2$, but the force? My instinct says it doesn't change overall, but I wanna be sure. Force is mass times acceleration. For a single molecule, it will impart the same force no matter where it hits. (Imagine drawing a pressure sphere around the gas, the molecules bounce off that wall, imparting a force.) Thus, the actual force imparted by the molecules is the same no matter where we draw our "boundary". (This is odd, perhaps look up more information about pressure). 

Regardless, $PV \propto \frac{1}{r^2}r^3 \propto r $. So as everything expands, PV goes up. Which means the temperature should also go up... hang on, what? 

Ah, right, this is assuming V is the entire sphere. In the situation of a planetary nebula, it's actually going to be a ring (or cocentric rings). For a thin shell, the "volume" scales like a two-dimensional object instead of a three-dimensional. Which means that PV is proportional to a constant, meaning we are in a "sensitive to conditions" regime. T could go up or down depending on the specifics of the situation. This does make some sense, it must be quite difficult to cool below 3K if this is the only place we see it. 

Now, what does this say about why Hydrogen doesn't solidify in space? Well, the temperature is going down of the entire cloud... but each individual molecule continues at the same speed and will impart the same energy if it smacks into something. Planetary nebula launches aren't *slow* by any means. They're probably just moving too fast to latch onto each other! This seems reasonable to me, but I'd like a way to quantify it... 

Ah, let us consider the mean free path. 

$$ l = \frac{1}{\sigma n} $$

the mean free path is dependent upon the cross-sectional area and the number density per unit volume. The cross sectional area for a hydrogen molecule is mostly constant, so it's really dependent upon number density, which is dropping dramatically as the gas expands more and more over a very wide area. Thus, the mean free path goes up *dramatically*. 

Let's look up the densest planetary nebula we know of, see what the mean free path is there. [2] (an annoying source to dig through due to the archive's methods) gives us a density range for planetary nebula with the largest of 1000 particles per cubic centimeter, or 1e10 per cubic meter. Now small molecules and atoms are on the order of an angstrom in radius, or 1e-10 meters. Squaring this to get a rough cross section, we get a very loose mean free path of 1e11 meters, or 1e8 km. (Which is absurdly large even compared to the values I saw of about 5800 km). REGARDLESS, when your mean free path is measured on the scale of thousands of kilometers or more, you aren't having molecules interact. So the very procedure that makes this place very cold ensures the solid cannot form since there isn't enough density of hydrogen. This explains why there are no solid hydrogen chunks in the Boomerang Nebula.

But wait, a new consideration: so the expansion cooling won't freeze things. But hydrogen comes together in nebula due to gravitational contraction all the time, it makes stars. Of course the reason this doesn't create solid hydrogen is because as they get compressed everything heats up, just the way it is. Any sort of compression is going to increase the temperature, no doubt. 

So that leaves the nebula that aren't compressing. Surely with enough *time*, hydrogen will start to solidify? Stellar nurseries get hot due to, y'know, the *stars*. (Goolge says Orion is around 10,000 K. Yeesh!) Most nebulae appear to be quite hot, it seems, or the temperature doesn't matter. 

Ohohohoho! FINALLY found something! We get led back to 'Oumuamua. There was a theory that it may have been a chunk of hydrogen ice [3]. However it appears that this theory was discredited [4] but this will most *certainly* tell me what I need to know about the formation of hydrogen ice objects.

[3] tells us that H2's sublimation temperature in space is 6K. Good, that actually gives us quite a range of temperatures. 3-6 is acutally significant.

[4] Claims that solid hydrogen objects cannot form at all due to *collisional heating*. 

Ohohohoho I appear to have actually stumbled upon an interesting question, as there is active research into solid hydrogen grains. Excelllllllent... let's dig in. Quite fun.

Hydrogen objects are unlikely due primarily to collisional heating, however [4] does point out that such objects could still form by phase transition as given in another paper. Yay! 

In conclusion, hydrogen ice can only form naturally in unusual situations, in the very centers of the densest, coldest, isolated nebulae. And there are some significant challenges to actually attaining this and it may not acutally occur.

Okay so now I want a good source for the ambient pressure of SPAAAACE. Ugh a good source is hard to come by for some reason... no matter, it can be calculated. PV=nRT. Let's just find the density of space. 

Interplanetary: 5-10 protons (might as well be hydrogen) per cubic centimeter. [6]

Interstellar: 1 hydrogen per cubic centimeter. [7].

Intergalactic: 1 atom (presumably hydrogen) per cubic meter. [8]. 

So we have densities here in number of molecules per a volume. Thus we can convert our ideal gas equation to

$$ P = \rho R T $$

The density should be in "per meters" so our values are 1, 1e6, 1e7. Note how *much* less dense the intergalactic medium is. However our density should also be in mols. Instead of adjusting these values though, we can just slap avogadro's number onto R. 

R = 8.314, the avocado number is 6.023e23, meaning our equation becomes

$$ P = \rho (1.38e-23) T $$

Now for T we can just use the T of whatever space we care about. Intergalactic medium will be 3K, interplanetary somewhere around 40K and up. So we get...

$$P_{interstellar} = 4.14e-23 \text{ Pa}$$
$$P_{interplanetary} = 5.52e-15 \text{ Pa}$$

Such a *wide* range of values. We can con convert this to bars by 1e-5 (which is also very close to atm for comparison's sake). 

$$P_{interstellar} = 4.14e-18 \text{ bar}$$
$$P_{interplanetary} = 5.52e-10 \text{ bar}$$

$$ U_r = \frac{3GM^2}{10R} $$

Energy available due to gravitational contraction of an object of uniform density. Basically none of these planets are of uniform density though, so this is gonna be one of those "maybe an order of magnitude correct, if that" calculations. This also won't help all that much with specific phase transitions as we can't just draw a line on the phase diagram--after all, there's the unique energy requirement to *execute* a phase transition. But we can still get a general idea of how pressure will go up with temperature. 

Ignoring phase transitions, temperature is linearly related to energy. But what of the pressure? Well, the pressure is going to scale with R for uniform density. (The core density anyway). Quick lookups indicate that for a sphere, core pressure goes with $R^2$. 

Unfortunately we currently have another parameter, M. But we happen to know M goes with $R^3$. Since M is squared, that means we get a whopping $R^5$ for energy and thus temperature dependence at the core. Meanwhile pressure is just $R^2$. Therefore, temperature is going to increase at a much faster rate than pressure in most cases... at the core. Since we're going to assume equilibrium for sanity's sake, the temperature must drop with the pressure toward the surface. 

Since temperature is going to increase so much more than pressure this keeps the result from being simple. Each individual planet will have its own phase diagram to go through. Some of them will hit liquid, others not, very hard to say. And this is ignoring radioactive heating and the like. Eventually, though, everything will cool to the "cold" state we were considering before. Most materials though probably can't "gassify" in the core. Though that would be weird...

We can, however, make some quantitative ideas for materials that follow the standard phase diagram plot. Which is most of these elemental substances, honestly, there are a handful of oddballs but they aren't obviously going backward. (Exceptions being Plutonium and Barium of all things.) For this, select a point on the diagram--this is the outermost layer of the planet. Then follow a trajectory of the cube root form since $R^2 / R^5 = 1/R^3$. Now there will be a lot of phase transitions in the way, of course. Regardless, this means that we can limit the kind of lines we draw: they have to be shallowly increasing in pressure but sharply in temperature. Interestingly on a log-log plot this appears linear, with a 1/3 slope. This line would showcase an extremely rough idea of planetary layers. However, this implies that a water planet should have a vapor layer, not a liquid one, under an initial shell in cold situations.
