# Mask Creator

## Create a mask that shows where all the lakes are on Titan.

### Within a resolution of 1 lat/lon point.

Okay so, trying to do this manually by inspeciton has proven difficult and full of decisions that are hard to make, visually it's a nightmare. However, one of our failed ways to validate the specular/lambertian Titan model can come in handy here: we can SCORE the various parts of an image based on whether they match the specular model well. While this was not a good way to validate, it WAS generally able to distinguish between lakes and the land next to them. (On images where it was already clean, to be fair). So we can use this to generate a mask!

First, let us create (and save) the mask itself.


```python
#Create the mask object and save it. NOTE: running this seciton of the code can overwrite the previous mask! Be very careful!

maskArray = [[0 for x in range(360)] for x in range(36)]
#36 lat points from 55 to 90
#361 lon points from 0 to 359. Rounding to 360 will go to 0.

with open('maskSeas.csv', 'w') as maskfile:
    for item in maskArray:
        outfile = ""
        for subitem in item:
            outfile = outfile + str(subitem) + ","
        outfile = outfile + "\n"
        #print(outfile)
        #maskfile.write(outfile)
```

First thing we need to do is read the file in. 

```python
import csv
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

maskArray = [[0 for x in range(360)] for x in range(36)]

with open('maskSeas.csv') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    i, j, k = 0, 0, 0
    for row in csv_reader:
        while (i < len(row)-1):
            maskArray[j][i] = float(row[i])+i
            i = i + 1
        i = 0
        j = j + 1
        line_count += 1
    
#Now DISPLAY!
#Display is in polar coordinates to give a good idea of what is actually going on.
#Polar assistance: https://stackoverflow.com/questions/73831636/pcolormesh-in-polar-coordinates-redux
fig, ax = plt.subplots(figsize = (9,9),subplot_kw=dict(projection='polar'))

#Now we do need to invert an axis.
#i=0
#while (i < len(compiledData)):
#    compiledData[i] = np.flip(compiledData[i])
#    i=i+1
#compiledData = np.flip(compiledData)

rs   = np.linspace(55,90,36) 
thetas = np.linspace(0,2.01*np.pi,360)
X,Y = np.meshgrid(thetas,rs)

image = ax.pcolormesh(X,Y,maskArray,cmap="seismic")
ax.set(ylim=(90,55))
fig.colorbar(image,ax=ax)
ax.set(title = 'Who knows what the plot will be?') #Remove ticks.

plt.show()
```

Now that we have our displays working, it's time to alter our mask. The plan is as follows.

1) read in cube data

2) run cube data through the "scoring" system against the specular data model.

3) set a threshold that will set some pixels to -1 and some to 1. Will change on a pixel by pixel basis.

4) use these values to determine what lats and lons in the iamge are lake and land.

5) adjust the model.
