Data from: https://ncses.nsf.gov/indicators/states/indicator/state-support-for-higher-education-per-fte-student

Let's see how fiscal spending on higher education has changed from 2000 to 2021. 

```python
import matplotlib.pyplot as plt

years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021]
munz = [7518, 7313, 7067, 6628, 6211, 6286, 6572, 6766, 6939, 6347, 5688, 5596, 5257, 5350, 5653, 5987, 6103, 6255, 6248, 6473, 6656, 6671]

fig, ax = plt.subplots()
ax.set_xlabel('Year')
ax.set_ylabel('Dollars')
ax.set_title('Higher education state support per full time student inflation adjusted')
ax.plot(years, munz, color='b')
ax.plot([2008,2008.0001], [5000,8000], color='r')
ax.plot([2014,2014.0001], [5000,8000], color='r')


# https://stackoverflow.com/questions/332289/how-do-i-change-the-size-of-figures-drawn-with-matplotlib 
# Setting size was annoying.
fig.set_size_inches(9,9)
fig.title = 'Title'
# ax.set_yscale("log") # Found in matplotlib's documentation.
# ax.legend()


# Bin values need to be normalized so the addition of all bins equals 1.
# then plot with matplotlib. Consider result. Make lots of plots and compare always.
```
