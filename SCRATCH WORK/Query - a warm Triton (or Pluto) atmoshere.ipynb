## A Proper Formulation of the Question

Triton, Pluto, and various large TNOs have vast quantities of frozen nitrogen on their surface, as well as some other gasses. Earth's and Titan's atmospheres are nitrogen-dominant, and so one wonders if the conditions on the frozen nitrogen worlds were warmed, if a substantial atmosphere would present itself. Both Pluto and Tritan have confirmed tenuous nitrogen atmospheres as it is, the question is how big would it get, and how would we compare it to Earth and Titan?

## Preliminaries: What will we compare it against?

The simplest comparison I could think of was simply to compare the masses of the atmospheres: specifically, the masses of the nitrogen contents. Let's begin with our known atmospheres (skipping Venus as it's not Nitrogen-dominated). 

Earth's atmosphere has a mass of 5.1480e18 kg, and has a 78% nitrogen content. However we do have to ask if this percentage is by mass or number. Looking it up, it appears the 78% value is a mole fraction value derived from the parts per million. So in order to get this by *mass* we need to compare it to the other constituents of the atmosphere. 

78% N2, 21% O2, and 1% Ar. Yes, the molecular forms are what's counted here. And we are rounding to the nearest percent which means trace gasses are irrelevant. (Even the nearest tenth of a percent woudln't pick up CO2, the next biggest constituent.) The amu measurements for these molecules are 28, 32, and 40 respectively. Multiplying the percents by the amus we get the relative mass numbers for each; 2184, 672, 40. The units are irrelevant as these are just going to be used for ratios. The one we care about, nitrogen, ends up with only a 75% composition by mass. 

This means that in Earth's atmosphere there are 3.861e18 kg of nitrogen. Good. 

Titan's composition is 94% nitrogen and 6% methane when rounded to the nearest percent. (Hydrogen, the most populous trace element, would appear if we went to a tenth of a percent.) Methane has a molecular mass of 16, and so with that we can find the nitrogen mass percentage of 96%. This is sensible as it is more massive per molecule than methane is, and notably is a very small adjustment compared to Earth's. (Notably the exact nature of Titan's atmosphere changes with elevation but we are only looking for bulk properties so this will do). Thus, all we need now is the mass of Titan's atmosphere. However *this* number is not just readily available in an obvious location, so we'll have to dig for it.

And we have run into a snag because the source wikipedia cites (Titan: Explorign an Earthlike World) is not publicly avalable, so I can't use that number. So let's try to work backward: how can we determine the mass of Titan? One factoid about the moon that *is* commonly known is that the surface pressure is around 1.5 bar. There are formulas for calculating surface pressure, as surface pressure is determined by the mass of air above something. P=F/A, pressure is force over area. Now everyone knows F=ma, force is mass times acceleration. Well, we want to find the mass, so we can make this substitution to get P = ma/A, or P/a = m/A. A is the "area" so we can just figure out the "mass per meter squared" and then extend that over Titan's surface area. 

a, the acceleration, is the acceleration due to gravity at Titan's surface. The clever among us will note that gravity changes as we go up from the moon, but the vast bulk of the atmosphere is close to Titan's surface and thus does not feel much difference in gravity. Titan's surface gravity is another readily available number, 1.352 m/$s^2$. Converting pressure to pascals, we now get m/A = 150000/1.352 = 110947 kg/$m^2$. 

That's a lot of mass and we're about to make it a lot larger via the surface area of Titan, with our friend $A = 4\pi r^2$. However, the radius for Titan is a little odd, are we talking visual radius or radius to the surface? We want surface radius, but we've got to be sure we get that number. The commonly cited value is 2575 km, and examining the examinations performed by [1] using radar, in which the atmosphere is invisible, this appears to be the one we're looking for. Using this we find that the *total* mass over all of Titan's surface area is... 9.244e18 kg. This is significantly larger than the value given in wiipedia of 1.19 Earth's, however it is still within the same order of magnitude and our final analysis does not require much precision, so we shall leave it as a comparison point. 

Anyway, for our purposes a "substantial atmosphere" refers to something on the order of 1e18 kg. (The mass of Titan's atmosphere should really be known far better than one order of magnitude but we'll leave that for now.)

Anyway, now we can attack the meat of the problem since we have a goalpost.



## How Much Nitrogen?

It's pretty well known that Pluto and Triton and other similar objects have vast frozen nitrogen shells all around them, though notably at differing layers of thickness and specificity. This is presumably where our atmosphere would come from if we heated up these worlds: the surface nitrogen sublimes and becomes the atmosphere. What we want to know is how much do we *get?* 

This essentially boils down to a problem of how much nitrogen is frozen on the surface of these worlds. Since Pluto has the most data on it due to New Horizons, we shall use Pluto as the prototype for a final answer. This is the part there's no clear, reliable data on, and it's time to jump through some hoops. 

The simplest way to judge the nitrogen content is by complete bulk composition, which is easy enough to find. Extrapolating from Cl Chondrites (a type of meteorite) we can apply their nitrogen abundance to Pluto and find that Pluto has 3e19 kg of nitrogen [2]. That's plenty, right? 

Well. Uh, no, becasue that includes all nitrogen, even nitrogen locked away beneath the surface that isn't going to release just by heating the dwarf planet up. Still, we only need to actually turn 1/30th of this into the atmosphere to hit our goal, so this is looking good but is not a satisfactory answer. What we really need is the *surface* content of Pluto's nitrogen. (Even an argument by nitrogen "floating to the top" doesn't work, as methane and water are both lighter than molecular nitrogen; though one can argue that Pluto must have been significantly warmer in the past and so then all the nitrogen was gaseous, but that nitrogen might have escaped and you see the issue here now right?)

So, in modern times, how much nitrogen ice is *currently* acessible from the surface? 

Very good question. We really don't know how far down the nitrogen goes. So my job is to find some way to constrain a ballpark estimate.

It just so happens that, on Pluto, the nitrogen ice has managed to fill the utterly massive Sputnik Planitia as a glacier of boggling proportions. At the very least this glacier will be evaporated, so we can use it as a very generous lower bound. [2] tells us Sputnik Planitia is 2.5-3 km deep, and since we're using a lower bound let's just let it be 2.5 km. The extent of the planitia is easy to find, 1050 by 800 km, for an extremely rough area estimate of 8.4e11 square meters of surface, so the volume is 2.1e15 cubic meters. So to find the mass of nitrogen we ignore contaminants (which, even if there are a lot, the rest of Pluto will definitely make up for it) and then multiply by the density of frozen nitrogen.

Except this is frozen nitrogen in a very exotic location so the density isn't just gonna be an easy look up. 

[3] tells us that the N2 ice is in the hexagonal crystalline beta-phase on Pluto. This would not be the case on Triton (as I recall from the Triton notebook) but we can still apply it as the overal nitrogen content shouldn't be vastly different. 

So, we now need to find the density of hexagonal crystalline beta-phase nitrogen. Fortunately Google stores a preview of [4] that gives us most of the information we need: a hexagonal close packing structure (hcp) with dimensions a=0.45nm and c=0.66nm. Of course we now ask if these are individual N atoms or N2 molecules doing the packing. [5] indicates that it's the molecules doing the packing, so every lattice point we pack has a full molecule. 

The volume of a single cell is that of a hexagonal prism with height c and side a, giving a volume of 0.347 cubic nm. Now, how many molecules are in an hcp cell? There's three interior ones, but then there's the ones on the faces that get "split up."



*HCP unit cell, from https://www.e-education.psu.edu/matse81/node/2134*

We can use the diagram to count. The face-central molecules are split in half so in total they are one molecule. The corners are split in half and *then* in thirds, making each one 1/6th of a molecule, so the corner ones all together add 2 more total molecules. 3+1+2 is 6, so we have 6 N2 molecules, or 12 N atoms, in a single cell. 

We can find the density from this. We want kg/m$^3$, but what we have right now is amu/nm$^3$ (168/0.347). We convert these values to 2.780e-25 / 3.47e-28 = 801.153 kg/m$^3$. A quick lookup claims normal nitrogen frost has a density of 850, which is remarkably close considering the dramatically different conditions. (So either we got lucky or there isn't much variation).

Anyway, with our density, we can find out how much nitrogen is stored up in our massive glacier. The result... 1.682e18 kg of nitrogen. 

In Sputnik Planitia. 

*There's a lot more nitrogen on Pluto's surface than that.*

So at the *lower* end of what we're trying to calculate here, the *lower* bound, we have *half* the mass of Earth's atmosphere. *half*. I'd consider that significant! Impressive! 

To get some idea of what we're dealing with ehre, the surface area of Pluto is 1.774e13 square meters, while Sputnik Planitia was just 8.4e11. There's enough surface area to hold 21 Sputnik Planitia's, and while the rest of Pluto is not so *concentrated* in Nitrogen, it should easily have enough to double what Sputnik Planitia has. Thus, we have reached an atmosphere with mass comparable to Earth's. 

While *lowballing*. 

I'd say that answers the question.
