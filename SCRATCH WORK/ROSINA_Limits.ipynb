# ROSINA Limits
## What We Can and Can't Do
### Gotta know all the limitations!

This notebook is a way to calculate various limitations of the ROSINA instrument, as outlined in https://core.ac.uk/download/pdf/33053779.pdf. This is intended for use in the NASA PSSS (Planetary Science Summer School), specifically on the ECHO Enceladus mission concept. 

First off, mass resolution changes with mass. Let's see that graph.

```python
import matplotlib.pyplot as plt
import numpy as np

# --- Data for the first function ---
# x ranges from 1 to 12
x1 = np.linspace(1, 12, 100)
# y = x / 500
y1 = x1 / 500

# --- Data for the second function ---
# x ranges from 12 to 150
x2 = np.linspace(12, 150, 200)
# y = x / 3000
y2 = x2 / 3000

# --- Data for the third function ---
# x ranges from 150 to 300
x3 = np.linspace(150, 300, 200)
# y = x / 500
y3 = x3 / 500

x4 = np.linspace(12, 150, 200)
# y = x / 500
y4 = x4 / 500

# --- Plotting ---
# Create a figure and an axes object
plt.figure(figsize=(10, 6))

# Plot each function on the same axes
plt.plot(x1, y1, label='RTOF', color='blue')
plt.plot(x2, y2, label='DFMS', color='green')
plt.plot(x3, y3, color='blue')
plt.plot(x4, y4, color='blue', linestyle='--')

# --- Adding plot details ---
# Add a title to the plot
plt.title('ROSINA instrument resolution over  full mass spectrum.')

# Add labels to the x and y axes
plt.xlabel('mass (amu)')
plt.ylabel('mass resolution (change in amu)')

# Add a grid for better readability
plt.grid(True)

# Add a legend to identify the lines
plt.legend()

# --- Display the plot ---
# Show the final plot
plt.show()
```

Lower mass resolution number is better. (Well, it's higher resolution, but the number is smaller as that means you can look in deeper. I love the english language.) DFMS is superior in the range it operates, which is not the very low end nor the higher mass end. This is what we use to determine how well we can differentiate molecules. 

But now, can we even detect molecules? The pressure ranges ROSINA can detect are 1e-15 for DFMS and 1e-17 for RTOF. Very low pressures, but the default pressure of the Enceladus plume is pretty low to begin with. 

From https://www.researchgate.net/publication/7250444_Cassini_Ion_and_Neutral_Mass_Spectrometer_Enceladus_Plume_Composition_and_Structure we have a range on the number density of the primary constituent (water) of 1e4 to 1e6 molecules per cubic centimeter. As our limits are measured in pressure, we have to calculate it. Fortunately ideal gas applies and so P=nkT. T is assumed to be 190K, the sublimation temperature of water. (what we expect it to be doing shooting out of that jet). 

So...

P = n(2.6232331e-21) = 2.6232331e-11 to 2.6232331e-9 Pa = 2.6232331e-13 to 2.6232331e-11 mbar.

Now the plume may be denser or less dense in places, but this we can work with. This does however make the worst case scenario for DFMS rather bleak: only things of 1/100 the prevalence of water could be detected. In the higher poritons, 1/10000, which is better. 

RTOF has much better ratio sof 1/10000 to 1/1000000, but its resolving power is not as good.

https://www.researchgate.net/publication/253800316_The_water_vapor_plumes_of_Enceladus extends the range of water density, we can how say 1e4 to 1e8 (with occasional approachings to 1e9). From https://pmc.ncbi.nlm.nih.gov/articles/PMC6278931/ we get that there are diferent flybys with different densities, E3 has the 1e8, and since it's possible to get ten times that, 1e9 is considered thet op. 1e4 appears to be largely the background level. (We will point out, however, that this closest approach only takes roughly 800 total seconds to elapse, this is not two hours, so the 1e4 value is far more likely for our high-mass-resolution instrument. There is some evidence of trail off at 1e6 though. We may also simply be able to travel through it slower.)

anyway 1e4 corresponds to 2.623e-13 mbar, but we can increase it all the way to 2.623e-8 mbar at the maximum density. 

At maximum density RTOF is sensitive to a whopping 9 orders of magnitude. Amazing! But in our worst case scenario we have only 2 orders, not great.

A source https://iopscience.iop.org/article/10.3847/PSJ/aced90#psjaced90t1 predicted the concentration of glycine to be 0.1 M, where M is molarity, or mols/L. Water itself has roughtly ~55 moles in a liter of itself, so glycine is predicted to be at relative concentrations of 0.0018. This is a brine calculation, so this is most assuredly enhanced. So enhanced glycine detection requires three orders of magnitude sensitivity.
