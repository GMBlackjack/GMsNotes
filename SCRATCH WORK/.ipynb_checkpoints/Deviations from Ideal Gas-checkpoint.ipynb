# Deviations From Ideal Gas

## "How Wrong Are We"

### You'd think this would be possible to do analytically but apparently not, at least not reasonably.

The problem: we want to know how closely Nitrogen behaves to an ideal gas at the conditions on Titan so we can know that our model isn't insane. The only problem is that measuring deviations from ideal gas isn't exactly *simple*. The general way it appears to be done in the literature is plot PV versus P. As for an ideal gas PV=nRT, then PV will be a constant at any given temperature as n and R are not changing.

One may ask: why do we not plot P versus T? Well, this is largely due to the fact that the most precise equation of state for a gas is the Virial Equation of State (stated below assuming n=1):

$$ P = \frac{RT}{V} \left( 1 + \frac{B(T)}{V} + \frac{C(T)}{V^2} + ...\right) $$

https://chem.libretexts.org/Bookshelves/Physical_and_Theoretical_Chemistry_Textbook_Maps/Physical_Chemistry_(Fleming)/02%3A_Gases/2.07%3A_Real_Gases

B and C are the second and third virial coefficients. (There is a D as well and we can go on forever but we generally don't need more than this). The problematic part is that B and C are temperature dependant and the relation between them is entirely emperical, so to actually plot any continuous lines for this equation of state you need to select a *specific* T and then plot the line. (Or you would need to interpolate. Which is suspect). 

So if you have to select T, that leaves P and V as your free variables. V is somewhat annoying to work with as we rarely have a specific volume in mind when working in things, like, atmospheres, so it is instead traditional to use PV to actually examine the differences between ideal gas. It's just that writing PV in terms of P for the virial equation of state would involve solving a polynomial of degree N, where N is the orders of accuracy desired. 

So instead we have been forced into this little box here where we have to solve it numerically. The procedure is as follows:

Take in program inputs of T, B, and C, which really aren't independent but will have to be entered differently as they are empericcally determined. Calculate a table of values for P(V). Convert to V(P), which is as trivial as just rearranging the lists. Multiply V(P) by P so we have VP(P), which is what we want to graph in the first place. P will be in bars, V in liters. 

Program should graph the ideal gas, first order correction, and second order correction. Here we go...

```python
import matplotlib.pyplot as plt
import numpy as np
import math

#Temperature in Kelvin:
T = 98.0

#Second Virial Coefficient in L/mol
# https://www.sciencedirect.com/science/article/pii/S0021961497902305?ref=pdf_download&fr=RR-2&rr=84a2386a3c4b6ce2
#Nitrogen at 98K
B = -0.1665
#Methane at 191K
#B = -0.11631
#Extrapolated methane.
#B = -0.310

#Third Virial Coefficient in (L/mol)^2
#Nitrogen at 98K
C = 0.0002330
#Methane at 191K
#C = 0.0004741 

# Use 0 if unknown.
# C = 0 

#van der waals numbers
# https://en.wikipedia.org/wiki/Van_der_Waals_constants_(data_page)
#Nitrogen
a = 1.370
b = 0.0387
#Methane
#a= 2.253
#b= 0.04278

R = 0.08206 

# set up V, which is our primary variable. This is our domain. 
V = np.linspace(0, 1000, num=100000)
V = np.delete(V,0) #remove the zero we don't want to cause a problem with the math.
P1 = (R*T)/V #The ideal gas law. 
P2 = P1*(1 + B/V)
P3 = P1*(1 + B/V + C/(V*V))


# Van Der Waals
P4 = (R*T)/(V-b) - a/(V*V)

fig, ax = plt.subplots()
ax.set_xlabel('P')
ax.set_ylabel('PV')
ax.set_title('Deviation from Ideal Gas Law')
ax.plot(P1, P1*V, color='r', label = "Ideal")
ax.plot(P2, P2*V, color='g')
ax.plot(P4, P4*V, color='purple', label="Van Der Waals")
ax.plot(P3, P3*V, color='b', label="Virial")


# https://stackoverflow.com/questions/332289/how-do-i-change-the-size-of-figures-drawn-with-matplotlib 
# Setting size was annoying.
fig.set_size_inches(9,9)
fig.title = 'Title'
plt.ylim(5,10)
plt.xlim(0.0,8)
# ax.set_yscale("log") # Found in matplotlib's documentation.
ax.legend()
```

And the FINAL step... find a percent difference at 1.5 atmospheres! Or bar. 



```python
fig, ax = plt.subplots()
ax.set_xlabel('P')
ax.set_ylabel('PV% Difference')
ax.set_title('Percent Difference PV from Ideal')
ax.plot(P2, ((P2*V)-(P1*V))/(P1*V)*100, color='g')
ax.plot(P4, ((P4*V)-(P1*V))/(P1*V)*100, color='purple', label="Van der Waals")
ax.plot(P3, ((P3*V)-(P1*V))/(P1*V)*100, color='b', label="Virial")


# https://stackoverflow.com/questions/332289/how-do-i-change-the-size-of-figures-drawn-with-matplotlib 
# Setting size was annoying.
fig.set_size_inches(9,9)
fig.title = 'Title'
plt.ylim(-10,0)
plt.xlim(1,2)
# ax.set_yscale("log") # Found in matplotlib's documentation.
ax.legend()
```

Conclusions:

Nitrogen at Titan conditions is only a 3% deviation. Not bad.

Methane at Titan conditions has no data at the right tempreatures, but using a very primitive extrapolation (i.e. eyeballing an exponential curve) we can get B = -0.310 ish. This results in a 6% difference at Titan conditions, which is significant but still not terrible. If we assume our extrapolation is terrible and go with -0.410 instead this increases to 8%.

Anyway I want to plot nitrogen at several temperatures now, which means we're going to have to normalize it somehow, force the Ideal Gas line to equal 1. We can actually do that by dividing out some things, and plotting PV/RT rather than just PV.

```python
import matplotlib.pyplot as plt
import numpy as np
import math

#Temperature in Kelvin:
T1 = 98.0

#Second Virial Coefficient in L/mol
# https://www.sciencedirect.com/science/article/pii/S0021961497902305?ref=pdf_download&fr=RR-2&rr=84a2386a3c4b6ce2
#Nitrogen at 98K
B1 = -0.1665

#Third Virial Coefficient in (L/mol)^2
#Nitrogen at 98K
C1 = 0.0002330

#van der waals numbers
# https://en.wikipedia.org/wiki/Van_der_Waals_constants_(data_page)
#Nitrogen
a = 1.370
b = 0.0387

R = 0.08206 

# set up V, which is our primary variable. This is our domain. 
V = np.linspace(0, 1000, num=5000)
V = np.delete(V,0) #remove the zero we don't want to cause a problem with the math.
PI = (R*T1)/V #The ideal gas law. 
P1 = ((R*T1)/V)*(1 + B1/V + C1/(V*V))
# Van Der Waals
PV = (R*T)/(V-b) - a/(V*V)

T2 = 150.0
B2 = -0.0718
C2 = 0.0002700
P2 = ((R*T2)/V)*(1 + B2/V + C2/(V*V))

T3 = 200.0
B3 = -0.0357
C3 = 0.0001935
P3 = ((R*T3)/V)*(1 + B3/V + C3/(V*V))

T4 = 300.0
B4 = -0.0061
C4 = 0.0001460
P4 = ((R*T4)/V)*(1 + B4/V + C4/(V*V))

T5 = 340.0
B5 = 0.0019
C5 = 0.0001400
P5 = ((R*T5)/V)*(1 + B5/V + C5/(V*V))

# wanted a really high temperature: https://web.mit.edu/8.13/8.13c/references-fall/aip/aip-handbook-section4i.pdf
T6 = 1400.0
B6 = 0.00335
P6 = ((R*T6)/V)*(1 + B6/V)

# found an 80 K https://aiche.onlinelibrary.wiley.com/doi/pdf/10.1002/aic.690100510
T7 = 80.0
B7 = -0.231
P7 = ((R*T7)/V)*(1 + B7/V)

fig, ax = plt.subplots()
ax.set_xlabel('P')
ax.set_ylabel('PV')
ax.set_title('Deviation from Ideal Gas Law')
ax.plot(PI, PI*V / (R*T1), color='r', label = "Ideal")
ax.plot(PV, PV*V / (R*T1), color='purple', label = "Van der Waals")
ax.plot(P7, P7*V / (R*T7), color='orange', label="80K")
ax.plot(P1, P1*V / (R*T1), color='b', label="98K")
ax.plot(P2, P2*V / (R*T2), color='pink', label="150K")
ax.plot(P3, P3*V / (R*T3), color='g', label="200K")
ax.plot(P4, P4*V / (R*T4), color='gold', label="300K")
ax.plot(P5, P5*V / (R*T5), color='olive', label="340K")
ax.plot(P6, P6*V / (R*T6), color='cyan', label="1400K")

# https://stackoverflow.com/questions/332289/how-do-i-change-the-size-of-figures-drawn-with-matplotlib 
# Setting size was annoying.
fig.set_size_inches(9,9)
fig.title = 'Title'
plt.ylim(0.8,1.1)
plt.xlim(1,2)
# ax.set_yscale("log") # Found in matplotlib's documentation.
ax.legend()
```
