# How do we Predict the Properties of Chemical Elements from Quantum Mechanics?

Really short answer: we use the Schrodinger Equation to calculate the arrangement of electrons in their orbital shells, and then use this determined arrangement to predict all the other properties of the atom.

Still short answer: In order to understand the properties of an atom we must first know its "shape" or "arrangement". This is a little bit of a nebulous concept in quantum mechanics as particles are in several places at once and yet nowhere, but there is still a higher chance of finding particles in particular places than others, thus defining a "shape" that the electron occupies. This shape is called an orbital, and an orbital cannot just take *any* shape. Quantum mechanics says that everything at very small scales is discretized, (or "quantized), and that includes the orbital shapes: only certain shapes are allowed to exist. We have, through experiment, discovered that these orbital shapes are determined by the Schrodinger Equation. (In fact, every particle in standard quantum mechanics is described by the Schrodinger Equation.) With a complete solution for an atom, the Schrodinger Equation will tell you exactly what orbital shapes are available for electrons, where they are located, and what their energies are.

The main problem is that the Schrodinger Equation is really complicated. It can in fact only be solved perfectly for the case of the hydrogen atom. However, it turns out that the hydrogen atom solution can be extended to other atoms, and with few exceptions the patterns continue unimpeded and result in the sometimes surprisingly organized properties of the periodic table. (Noble gasses don't react, halogens react with *lots of things*, etc). 

Once the shape of the electron orbitals and their energies are known, it is possible to determine atomic properties. This is because almost all interactions atoms have with the outside world is due entirely to the behavior of their electron cloud, and as the cloud layers are quantized every atom of the same time (ignoring isotopes) will have the exact same cloud arrangement, the properties are consistent across hundreds of thousands of these atoms. How tight or loosely electrons are bounded determines reactivity, emission spectrum, color, and even phase transitions, among other things.  

The situation is not completely solved, as these predictions only work for the standard elements under normal conditions. Fortunately, all atoms that standard matter is made out of are considered "standard elements," the predictions don't really start to break down until you get past element 100, and elements that large only exist when humans are doing nuclear experiments or stars are exploding. And "normal conditions" is actually a wide range around "standard temperature pressure", or room temperature. Unless you start getting to near-vacuum, near-isolation, or near-reletivistic situations, everything still holds. 

Now for the *longer* answer.

# The Longer Answer

To determine properties of chemical elements, we have two major steps: 1) determine the electron orbital configuration and energies for that atom. 2) determine the atomic properties from the orbital configuration and entities. So, first, step 1. 

## 1) Determining Electron Orbitals

At the very start of the process we have to accept the existence of a few quantum mechanical things as though they were just proven: the properties of the electron and the proton. Now there is a very complicated theory behind how the quarks in a proton *make* the proton, but that's irrelevant for us. To us, a proton is just a positively charged particle that obeys the wave-particle duality of quantum mechanics just like everything else. The electron is treated the same way, a negatively charged particle subject to wave-particle duality. 

So we ask: once an electron is in orbit around a proton, what does the Schrodinger Equation say about it?

The time-independent Schrodinger Equation in three dimensions is the following:



Which is definitely a mess of mathematical jargon that we won't be getting too involved with. The important thing to know is what our inputs are: E is the energy level of the system, a number. U is the potential energy distribution of the system, which in this case is simply the attractive force between a positively and negatively charged particle. The symbol $\nabla$ is not a quantity but an operator, describing a sort of three-dimensional double derivative. $\Psi$ itself is actually the quantity we want to *find*: the wavefunciton, which determines the probability distribution for the particle. The other symbols at the far left can just be thought of as numbers for our purposes. 

When we solve the Schrodinger Equation, we usually find that there are a specific set of solutions that exist, a specific set of $\Psi$ equations that work, and each one will only have a specific energy value that allows for it. For the hydrogen atom system, there *are* a specific set of energies and wavefunctions that work, these are the only ones that *can* exist.

*How* we solve this equation involves a lot of mathematical trickery and clever workarounds, since we're not solving for a number but an entire three-dimensional equation $\Psi$. This is very difficult but it was one of the early quantum mechanical calculations, and still the only physical system that can be calculated out perfectly. This mess of math fully describes all possible orbital configurations an electron can take in a hydrogen atom.

We generally ignore the mathematical details of the interior of the solution unless we actually want to calculate something specific--the important takeaway is that the solution $\Psi$ for all space depends on three values: n, l, and m. These are three of the four so-called "quantum numbers." There are many illustrations given as to what these numbers physically represent, but I like to think of them like so: n is the "size" of the orbital, l is the "shape" of the orbital, and m is the "orientation" of the orbital. The three numbers are related in that they restrict the values the others can take. n can be any integer 1 or larger (though really large numbers never occur). l can take any integer value from 0 up to n-1. m can take the same values as l, except it can also take negative versions of those values. 

Thus we have a finite combination of orbitals. Here are some of the simpler shapes:



This image shows the orbital options for n=2. The top row is for l=0, in which case m=0 as well. The bottom row is l=1, creating a double-lobed shape. The different orientations are chosen by m: -1, 0, or 1. 

Each of these orbitals is a shape the electron cloud takes. The "surface" seen in the above image is the most probable location to find an electron if you force the wavefunction to collapse, but most often around an atom the electrons simply exist as these clouds of potential. 

Since we solved the Schrodinger Equation, we also know the energy for every one of these states. For instance, the lowest energy possible for hydrogen is -13.6 eV, which means it takes 13.6 eV (electron volts) to remove the electron from its bound state in the atom, freeing it. This procedure is called ionization. Lower energies have the potential to excite the electrom from the ground-state orbital into another orbital. 

So that solves the hydrogen atom. For hydrogen we can now calculate its properties, including its tendency to bond to itself. But we can't get into that yet, as we still have all the *other* types of atoms to consider. See, one thing I didn't mention above is one of the assumptions that was requried to get the hydrogen atom solution: that the situation is spherically symmetric. For a single electron around a single proton, this is true, the potential of the electron only depends on how far from the proton it is, not which direction it is pointing. But with more protons and electrons this is not the case. However, it turns out that the nucleus of the atom is so small compared to the size of the orbitals that for the vast majority of elements, it can be modeled as a "blob" of some positive charge and the effect on calcluation is neglegible. (This will cause another problem later, be warned.) 

Since we can model the atomic nucleus as a single "blob" without worrying about its structure, all we have to worry about are the electrons around the nucleus interacting with each other. This, unfortunately, cannot be exactly solved, but it is the kind of problem you can give to a computer and get results as precise as you want with iterative techniques. And it turns out that the orbital *shapes* for hydrogen do not change, even for very large atoms. The *energies*, however, do change, and change significantly, due both to the larger potential of the nucleus and electrons interacting with each other. But we can still calculate these out given enough computational resources. 

Knowing the energies of the various orbitals for any atom allows us to predict which orbitals are filled by electrons for neutral atoms that aren't currently excited by some outside force, which is the default state in most situations. The lowest energy orbitals will be filled first, followed by higher energy ones. Only two electrons can fit in a single orbital because quantum mechanics forbids two electrons from having the exact same energy state in the same physical location. (Two electrons can cram into a location since a spin-up and a spin-down electron do have slightly different energies. The spin number is the fourth so-called quantum number after n, l, and m). 

Different atoms fill up the electron shells in different orders due to the solved energy states being different. Most elements follow a simple predictable pattern in the quantum numbers, and in high school we learn this predictable pattern for filling up electron shells. However, this pattern is wrong, as larger elements can start shifting energy levels around--the Schrodinger Equation accoutns for this, so its solutions are correct, but it seems random when just examined. The procedure of solving the Schrodinger Equation, however, is the same for all of them. Larger elements start throwing wrenches into this picture, but for any small element we should be good. 

So, with the Schrodinger Equation solved or approximately solved for all atoms, we now have electron orbital shapes and energies. This is sufficient to start predicting propeties. How? Well...

## 2) Predicting Properties

The properties easily predicted from electron shell configurations are the same properties that have general trends across the periodic table.



The above image is by no means exhaustive. 

Two of these are derived directly from quantum mechanics: atomic radius is directly related to the size of the largest occupied orbital, and ionization energy is directly related to the energy level of the outermost electron. Now while we can just say these properties are the way they are because "The Schrodinger Equation said so" there is some physical intuition here. For size, we see that adding an electron and a proton to an atom usually makes it smaller, but adding lots of electrons and protons will make it larger. This is because the more electrons there are in a single "shell" of orbitals, the more tightly bound the orbitals become, getting closer to the nucleus. However, if a shell is completely full and a proton and electron are added, an entirely new shell needs to start filling up, and so the jump in size is tremendous. Ionization energy follows the reverse pattern: the closer to filled the outermost shell of orbitals is, the harder it is to remove an electron from an atom. Thus, by solving the Schrodinger equation, you automatically get size and ionization energy, almost directly. 

Electron affinity is the primary measure of how reactive an atom is. Atoms with high electron affinity "want" to grab extra electrons, while low affinity "wants" to give up electrons. This is also ultimately an effect of the energy and size of the outermost orbitals and their energies. Atoms with very few outermost electrons generally don't hold onto them very tightly, so only a little energy is needed to remove them; thus they have low electron affinity and are more likely to give electrons up to other atoms. Meanwile, atoms with nearly full (but not full!) outermost shells would like to *fill* those shells and thus arrive at a more stable configuration (which ultimately means the entire system is in a lower energy state in the end). Thus, they have a high electron affinity, and are very likely to grab electrons from elsewhere. Elements with a full electron shell are noble gasses, and they break the rules by neither wanting to give up nor to take electrons: they are stable. 

Electron affinity also changes with overall atom size. Very large atoms cannot hold onto their electrons very well as the outermost ones are far from the nucleus, and so their electron affinity is low. Small atoms meanwhile hold onto their electrons very tightly and find it easier to grab onto exterior electrons. Fluorine, the lightest atom with a 7/8 electron outer shell, is one of the most reactive things in existence. 

Metallic character is also explained similarly. Large atoms with few electrons are easy to coax into a state where their outermost electrons can flow from one atom to another. With a lot of atoms that share this property, you get a conductive metal. However, atoms with more filled outer shells are resistant to letting their electrons flow, and so are nonmetallic.

There are other properties we can determine from electron shell arrangement alone. The emission spectrum of an atom is determined by what possible photons can be emitted by electrons jumping to and from energy levels. Each energy level is specifically defined, so when an electron jumps from one to another, it has to either absorb or release a very specific amount of energy. That specific amount of release comes in the form of photons of particular colors. (Fun fact: this is one of the tests as to how accurate our predictive models are. We can predict with great accuracy what these wavelengths are going to be, and even the minute difference in energy levels between two electrons with different spin states.)

These can also go on to describe natural color of materials (energy levels are open to absorb certain wavelengths of light and reflect others. Most metals reflect most everything, but some like gold will absorb colors and leave golden light behind). Or we can explain phase changes (the outermost electron shell behavior determines how strongly atoms bond to each other) etc. etc. The exact procedures to do these predictions are a mess, but I have an example of the *result* of one of these calculations for silicon. 



This is an experimental and theoretical model of two bonded silicon atoms. The measurement is the valance electronic charge density: that is, the density of where the valence electrons are most likely to be. From calculations of this sort we start to predict things like what compounds the atom can form, what the crystal structure will be, etc. We've gotten so good at this that many materials are predicted by theory before they are synthesized in a lab. With all these tools, the basic properties of normal materials can be determined. What sorts of things will they bond with? How do they interact with matter? Etc. etc. etc. There does not appear to be a *single* method on how to do this, but rather several different methods and ways to build up to other properties, mostly restricted by how strong of a computer and how much time you have on hand.

However, we have left one thing unaccounted for. Radioactivity. Unfortunately, the procedures described above do not account for standard radioactivity (Technetium excepted??). See, radioactivity depends on the processes happening inside the nucleus of an atom in most cases. We approximated the nucleus of an atom as a positively charged blob earlier. In order to predict radioactivity, we will have to remove this assumption. And that makes the problem a much, much harder one. And yet, we have done it, and done it for extremely large atoms that have *never* been made in accelerators. So, how?
