# PART I: Fundamentals and Another Look at Complex Numbers

https://youtu.be/PNlgMPzj-7Q?si=jZKcLzP2Sefu_rdi

Declare a 2D vector space with the usual perpendicular unit vectors, but call those vectors $e_1$ and $e_2$ rather than using the usual hats, even though this is just x and y. (Note that it doesn't have to be though).

We define minus signs as the opposite directions of the unit vectors, and every vector in this space is $ae_1 + be_2$. And with that we have declared our space. 

DOT PRODUCT $u \cdot v = u_1v_1 + u_2v_2$. For orthogonal, the dot product goes to zero. Commutative. And self dot product is the magnitude.

WEDGE PRODUCT $u \wedge v$, creates a paralellogram with circulation (bivector). The circulation is from u to v. 

$ u \wedge u = 0; u \wedge v = - v \wedge u $ (anticommutative). Opposite circulation is considered negative. 

To actually figure out what object we get, we can use these properties we determined geometrically to get a result.

$$ u = u_1e_1 + u_2e_2 $$
$$ v = v_1e_1 + v_2e_2 $$
$$ u \wedge v = (u_1e_1 + u_2e_2) \wedge (v_1e_1 + v_2e_2) $$

We note that while it wasn't exatly specified before, we can still distribute this out like we were FOILing it, or distributing. Let's just assume it works for now even though it isn't obvious it would with the rules we've set out. (However scalars moving in and out is obvious.) 

$$ = u_1v_1(e_1 \wedge e_1) + u_1v_2(e_1 \wedge e_2) + u_2v_1(e_2 \wedge e_1) + u_2v_2(e_2 \wedge e_2) $$

Wedge product with itself is zero, and by flipping the product on the second of the remaining terms we arrive at

$$ = (u_1v_2 - u_2v_1) (e_1 \wedge e_2) $$

Notably that this is the *determinant*. This is equal to the area of the bivector paralellogram. The magnitude is also the cross product. 

SO what if we just ADDED the two parts together? Thus...

$$ uv = u \cdot v + u \wedge v $$

This is the odd step that you think wouldn't help but TOTALLY DOES. 

$uu = |u|^2$

For perpendicular:

$uv = u \wedge v$

In general the geoproduct is not commutative or anticommutative. 

We feed in two 1D objects nad get a 0D and a 2D object out. 

Using our generic vectors again, what does this all expand into?

$$ uv = (u_1v_1 + u_2v_2) + (u_1v_2 - u_2v_1) (e_1 \wedge e_2) $$

Now what if we apply this to the unit vectors directly?

$$ e_1e_1 = e_1 \cdot e_1 + e_1 \wedge e_1 = e_1 \cdot e_1 = 1 $$

$$ e_1e_2 = e_1 \cdot e_2 + e_1 \wedge e_2 = 0 + e_1 \wedge e_2 $$

So $e_1e_2 = -e_2e_1$ perhaps unsurprisingly.

But now... what of $(e_1e_2)^2$?

$$ e_1e_2e_1e_2 = -e_1e_2e_2e_1 = -e_1e_1 = -1 $$

Ah yes the properties if i!

Anyway the final result we get is scalars, two vectors, and the bivector "i" or "I". So 2D geometric algebra is 4D, as you write out the multivectors as {scalar, $e_1$, $e_2$, I}.

Back up and consider imaginary numbers as we know them for a minute.

$$ e^{\theta i} (a+bi) = acos\theta - bsin\theta + (asin\theta + bcos\theta)i $$

This is a rotation. However, while we often think of complex numbers as a+bi, we can also think of them as a+b$e_1e_2$, where is now a bivector! For those of us who like abstract algebra, this is an isomorphism. 

Thus we can do rotations in 2D geometric algebra like so. Take a vector u which is $ae_1 + be_2$, then do ui. 

$$ (ae_1 + be_2)e_1e_2 = ae_1e_1e_2 + be_2e_1e_2 = ae_2 - be_2e_2e_1 = ae_2 - be_1 $$

Which is the same as having rotated the vector u 90 degrees counterclockwise. Note that we had to multiply on the right. Multiplying on the left gives a different result:

$$ e_1e_2(ae_1 + be_2) = ae_1e_2e_1 + be_1e_2e_2 = - ae_1e_1e_2 + be_1 = - ae_2 + be_1 $$

It's the rotation in the other direction. 

We can ask "what would give us an arbitrary angle rotation?" Well we know that imaginary numbers are isomorhic, so how do we rotate imaginary numbers?

$$ e^{\theta i} (a+bi) = (cos\theta + sin\theta i) (a+bi) = acos\theta - bsin\theta + (asin\theta + bcos\theta)i $$

Which as it turns out does map directly to the geometric algebra space. 

$$ (cos\theta + sin\theta e_1e_2)(ae_1 + be_2) = cos\theta ae_1 +cos\theta b e_2 + sin\theta a e_1e_2e_1 + sin\theta b e_1 e_2 e_2 = cos\theta ae_1 +cos\theta b e_2 - sin\theta a e_2 + sin\theta b e_1 = (cos\theta a + sin\theta b)e_1 + (cos\theta b -sin\theta a)e_2$$

Doing it the other way rotates the other way.

$$ (ae_1 + be_2)(cos\theta + sin\theta e_1e_2) = cos\theta ae_1 +cos\theta b e_2 + sin\theta a e_1e_1e_2 + sin\theta b e_2 e_1 e_2 = cos\theta ae_1 +cos\theta b e_2 + sin\theta a e_2 - sin\theta b e_1 = (cos\theta a - sin\theta b)e_1 + (cos\theta b + sin\theta a)e_2$$

Of course we can still choose to write this in an exponential form, $e^{\theta e_1e_2}$ if we wish. Using "i" or "I" as desired as well. Note that this means the rotation is given by a bivector, specifically the unit bivector in 2D space. This makes sense as a bivector is an area with a *circulation*, naturally that circulation could introduce actual rotations.

So here we have A=B+C. We can with some finagling show:

$$ B \wedge A = (A-C) \wedge A = A\wedge A - C \wedge A = -C \wedge A = A \wedge C $$

So $B \wedge A = A \wedge C$. Interesting. Geometrically this says the extended paralellograms from B and C have the same area. This essentially IS the law of sines, though if we want it in the more familiar form:

$$ B \wedge A = A \wedge C \Rightarrow |B||A|sin\gamma e_1e_2 = |A||C|sin\beta e_1e_2 $$
$$ \Rightarrow basin\gamma = acsin\beta \Rightarrow bsin\gamma = csin\beta$$
$$ \Rightarrow \frac{sin\beta}{b} = \frac{sin\gamma}{c} $$

You can also extend this to show that it's also equal to $\frac{sin\alpha}{a}$. 

For the law of cosines, let's start with the expression for C.

$$ C = A-B \Rightarrow C^2 = (A-B)^2 \Rightarrow C^2 = A^2 + B^2 - AB - BA $$
$$ \Rightarrow C^2 = A^2 + B^2 - 2A\cdot B \Rightarrow c^2 = a^2 + b^2 - 2abcos\gamma $$

Thus the law of cosines again. We will be discovering that so, so much we take for granted can be derived *from* geometric algebra. 

# PARV IV: Linear Algebra and Cramer's Rule

https://youtu.be/dnzUgDl43rQ?si=UK4fLzCOq1MvxAbA

What else can we get from this? 

Well say we are solving a linear combination equation.

$$ a_1x + b_1y = c_1 $$
$$ a_2x + b_2y = c_2 $$

Which we can write with vectors as 

$$ x\vec a + y \vec b = \vec c $$

So, how to we solve for x and y then with geometric algebra? Well, we wedge it.

$$ x\vec a \wedge \vec b + y \vec b \wedge \vec b = \vec c \wedge \vec b $$
$$ \Rightarrow x\vec a \wedge \vec b = \vec c \wedge \vec b $$

Of course solving for x here appears to need an inverse... oh or we can just divide (assuming the result isn't zero).

$$ \Rightarrow x = \frac{\vec c \wedge \vec b}{\vec a \wedge \vec b} $$

We can do this since the division is over a bivector, so really what we end up with is the "area ratio" between the two, leaving a sclaar. Also known as UNITS WORK OUT! 

Similarly for y we get $y = \frac{\vec a \wedge \vec c}{\vec a \wedge \vec b}$

This is all entirely arbitrary. Neat!

Let's work out an arbitrary wedge product... but do it from the matrix form of linear combinations so we can find something new out.

$$ A = \begin{bmatrix} 
a_1 & b_1 \\
a_2 & b_2 \end{bmatrix} $$

Let's do the wedge product between a and b.

$$ \vec a \wedge \vec b = (a_1e_1 + a_2e_2)\wedge(b_1e_1 + b_2e_2) = a_1b_2(e_1\wedge e_2) + a_2b_1(e_2 \wedge e_1)$$
$$ = (a_1b_2 - b_1a_2) (e_1 \wedge e_2) = det[A] e_1e_2 $$

This is the familiar 2x2 matrix determinant. Which is usually just some mysterious number but now we have a way to derive it. (But what of higher order determinants?) 

If we replace wedge products with determinants we get Cramer's rule. In short cramer's rule is

$$ x_i = \frac{det[A_i]}{det[A]} $$

The specifics of how this work is irrelevant to our purposes, as it involves some linear algebra. Our victory here is that we can get the same result from geometric albegra without finagling with the normal proof for Cramer's rule. 

A note on the determinant = 0 case. Since the determinant is an outer product, and an outer product is the paralellogram area, that implies that a det=0 case would be between two vectors that are parallel and thus have no wedge product. Visualization power, activate! 

# PART V: Two Reflecitons is a Rotation

https://youtu.be/Hy2gbdbrJZ8?si=jMkZXjQS9OBVkiUY

Okay so you know that challenge problem that was done? Yeah, we're returning to it, this video basically gives the solution. 

$$ w^{-1}v^{-1}uvw $$

Now earlier we showed that this is also equivalent to a single reflection. The goal is to show that this is ALSO equivalent to a single rotation. 

Now this makes sense: as long as a line goes through the origin (which it has to if we're reflecting across a vector) then the reflection itself is just a rotation. This *logically* makes sense. It also does mathematically to: any vector of the same magnitude as a nother can be transformed to it via a rotation, no exceptions. Refleciton does NOT adjust magnitude. So reflections are rotations and rotaitons are reflections, and in 2D space this is always true for *vectors*. 

Now the reason this realization is so useful is because it allows us to use double reflections to create a rather intuitive rotation formula that will apply to higher dimensions. We also note that our proof of the inversing did in fact come in handy, heh.

$$ w^{-1}v^{-1}uvw = (vw)^{-1}uvw = \frac{1}{|v||w|}e^{-\theta e_1e_2} u |v||w| e^{\theta e_1e_2} = e^{-\theta e_1e_2} u e^{\theta e_1e_2} $$

Behold the rotaiton SANDVICH.

This is in fact superior to our previous answer not because it's more correct--ours was still right--but this result is a specific usable result. We stated that a single reflection existed, NOT what it actually WAS. Here though we have a specific rotaiton that does the double reflection. Furthermore we can actually...

$$ e^{-\theta e_1e_2} u e^{\theta e_1e_2} = u e^{\theta e_1e_2} e^{\theta e_1e_2} = u e^{2\theta e_1e_2} $$

Which is to say the double refelction is a double rotation through the angle between the two reflection vectors. Note that if the reflection vectors are the same, they have zero angle, and the double reflection results in no movement! In essence this is merely more useful. Naturally a double rotation can also be accomplished with a reflection of some kind.

# PART VI: Kepler Problem 1 

## Oh no we're gonna be here a while...

https://youtu.be/VApO4qUgAYY?si=yPs58UssWnoctIwM

Also known as the Two-Body problem! Let's solve it with geometric algebra!



The situation, as we have it, is gravity. We have some relations here:

$$ \vec r = \vec r_2 - \vec r_1 $$

Due to LaTeX being annoying we can't use both vector symbols and dots, so just remember that quantities here are vectors when dotted.

And from F=ma we can produce:

$$ \ddot r_1 = \vec F_{12} / m_1 $$
$$ \ddot r_2 = \vec F_{21} / m_2 $$

And from Newton's Third Law:

$$ \vec F_{12} = -\vec F_{21} $$

So we can say

$$ \ddot r_1 = \vec F_{12} / m_1 $$
$$ \ddot r_2 = - \vec F_{12} / m_2 $$ 

Define $\hat r$ to be the unit vector of $\vec r$. With the law of gravity we use

$$ \vec F_{12} = G \frac{m_1m_2}{r^2}\hat r $$

And thus

$$ \ddot r_1 = G \frac{m_2}{r^2}\hat r $$
$$ \ddot r_2 = - G \frac{m_1}{r^2}\hat r $$ 

And also by taking derivatives

$$ \ddot r = \ddot r_2 - \ddot r_1 $$
$$ \Rightarrow \ddot r = - \frac{G(m_1+m_2)}{r^2} \hat r $$
$$ \Rightarrow \ddot r = - \frac{\mu}{r^2} \hat r $$

Where $\mu$ is just there for bookkeeping and is known as the gravitational parameter. 

There has been no geometric algebra yet. This is just the radial component of our problem.

We'll also be needing the angular momentum:

$$ \vec L = \vec r \times \vec p $$

However the cross product is really annoying and is inherently three dimensional. What if we really do wish to restrict ourselves to 2D? Well, we can do that! With GEOMETRIC ALGEBRA! You can probably see how one might represent the cross product using bivectors if you are familiar with the cross product. Otherwise just strap in and let's go!

So, define our coordinates. Let $\vec r$ be the vector from the pivot, and $\dot r$ be the direction of motion. We can create a bivector from this:

$$ H = \vec r \wedge \dot r $$. Naturally this does, in fact, circulate in the direction of rotation. For circular motion it's also easy to see that nothing particuarly changes here. However, now we need to take the derivative of a wedge product. One might think it odd but the product rule still applies here. 

$$ \dot H = \dot r \wedge \dot r + \vec r \wedge \ddot r = \vec r \wedge \ddot r $$ 

Now, remember what we showed in the previous part about $\ddot r$ pointing in the -r direction? Well paralell wedge products are zero. This means..

$$ \dot H = 0 = \frac{d}{dt} (\vec r \wedge \dot r) $$

This "the time derivative is zero" is the definition of a CONSERVED QUANTITY! Man that was easy. 

Now let's move to Kepler's Second Law. (don't think too hard about the order). 

So, Kepler's Second Law is about the area swept out at any given moment by an orbiting body. Well, we can represent the motion of an object with the position vector and tis velocity vector, so we can construct a paralellogram $\vec r \wedge \dot r dt$. Of course the actual area swept out is a triangle so we divide thsi by two. Thus:

$$ dA = \frac12 \vec r \wedge \dot r dt $$
$$ \dot A = \frac12 \vec r \wedge \dot r $$
$$ \dot A = \frac12 H$$

So a nice result. However we do need to point out that we pulled some weird derivative jujitsu there. Careful when dividing by differentials. 

The other point of this result is that since H is conserved, that means the rate of area swept out is constant. Which is Kepler's Second Law. Yay!

Okay so now let's look at some weird unusual things. First of all, show that $\vec r \cdot \dot r$ is zero. We do this by pulling *out* a time derivative, doing the product rule in reverse.

$$ = \frac12 \frac{d}{dt}(\vec r \cdot \vec r) $$

But wait, a dot product... is a constant. And a constant has a time derivative of zero. Thus, the dot product of a vector with its "velocity" vector is zero. This would appear to be absolutely general. (The video uses unit vectors for this but changing the magnitude will have absolutely no effect on this). 

Anyway, that's useful, and so let's use it to figure out more about H.

$$ H = \vec r \wedge \dot r = (r \hat r) \wedge (\dot r \hat r + r \frac{d}{dt} \hat r) $$

I will point out that I would love to put a derivative, hat, and vector on multiple things, but LaTeX wont' let me. So we now have notation confusion abounding. Lovely. Anyway, regardless of all that, what we've done is split the components up into their individual parts, reminding everyone that $\dot r$ as a vector would actually have multiple components and be subject to the product rule. Anyway distribute through the wedge product now--the $\hat r$ terms cancel due to being the same and we are left with

$$ = r\hat r \wedge r \frac{d}{dt} \hat r = r^2 \hat r \wedge \frac{d}{dt} \hat r = r^2 (\hat r \frac{d}{dt} \hat r - \hat r \cdot \frac{d}{dt} \hat r )$$

We previously showed that the dot product goes to nothing. And so...

$$ H = r^2 \hat r \partial_t \hat r $$

Where we introduce the relativistic version of the time derivative operator to keep us all sane. For the record this is the angular momentum per unit mass equation.

First of all just note that $\hat r \cdot (\dot r \hat r + r \partial_t \hat r) = \dot r \hat r \cdot \hat r = \dot r$. Useful.

Anyway let's return to our vector equation and work with it a bit to find more stuff. We use the above thing in the third line. 

$$ \ddot r = \frac{-\mu}{r^2} \hat r $$
$$ \Rightarrow \ddot r \cdot \dot r = \frac{-\mu}{r^2} \hat r \cdot \dot r $$
$$ \Rightarrow \frac12 \partial_t (\dot r \cdot \dot r) = \frac{-\mu}{r^2} \dot r $$
$$ \Rightarrow \frac12 \partial_t (v^2) = \partial_t \frac{\mu}{r} $$
$$ \Rightarrow \partial_t (\frac12 v^2) = \partial_t \frac{\mu}{r} $$
$$ \Rightarrow \partial_t (\frac12 v^2 - \frac{\mu}{r}) = 0 $$

More conserved quantities! Always good. This turns out to literally be the Energy E. (Total energy per unit mass, really). It's easily visible as a kinetic energy term and potential energy term. E = T + U. If we're particuarly clever we can also note that this is the Hamiltonian. Thus both energy and angular momentum are conserved!

# PART VII: Kepler Problem 2

## Time for magic!

For no particular reason, let's...

$$ H \ddot r = r^2 \hat r \partial_t \hat r \frac{-\mu}{r^2}\hat r $$
$$ = -\mu \hat r \partial_t \hat r \hat r $$
$$ = \mu \partial_t \hat r \hat r^2 $$
$$ = \mu \partial_t \hat r $$

And so...

$$ H\ddot r - \mu \partial_t \hat r = 0 $$
$$ \partial_t(H\dot r - \mu \hat r) = 0 $$

The derivative can be extracted as both H and $\mu$ are constants. This is a new conservation law. 

$$ H\dot r - \mu \hat r = \vec c $$

Note that H is a bivector, so the first term is a rotated vector, and the other is a scaled vector. $\vec c$ is known as the Laplace vector. The curious thing is that the Laplace vector is entirely conserved--it can't even rotate, it must always point the exact same direction. This is knowing full well that the radial direction and velocity direction change. 

So, starting with this new mysterious conserved quantity, we start finagling:

$$ H \dot r = \vec c + \mu \hat r $$
$$ \Rightarrow H \dot r\vec r = \vec c\vec r + \mu \hat r\vec r $$
$$ \Rightarrow H (\dot r\cdot \vec r + \dot r\wedge \vec r ) = \vec c \cdot \vec r + \vec c \wedge \vec r + \mu \hat r \cdot \vec r $$
$$ \Rightarrow (\dot r\cdot \vec r)H - H(\vec r\wedge \dot r) = crcos\theta + \vec c \wedge \vec r + \mu r $$
$$ \Rightarrow (\dot r\cdot \vec r)H -H^2 = crcos\theta + \vec c \wedge \vec r + \mu r $$

At this poitn we note that we have bivector and scalar sections, so we can split this off into two different equations.

$$ (\dot r\cdot \vec r)H = \vec c \wedge \vec r $$ 
$$ -H^2 = crcos\theta + \mu r $$

"Wait, $H^2$ is a scalar?" Yes, remember, the bivector squared is -1. Usually this quantity is superceded using $-H^2 = h^2$ to avoid leaving unnatural-looking signs.

$$ h^2 = r(ccos\theta + \mu) $$
$$ r = \frac{h^2}{\mu + ccos\theta} $$

Neat. Usually though we want to write it this way

$$ r = \frac{h^2/\mu}{1 + (c/\mu)cos\theta} $$

For notational reasons. Usually $p = h^2/\mu$ and $e = c/\mu$, so...

$$ r = \frac{p}{1 + ecos\theta} $$

You might recognize e: it's the eccentricity. p is the semi-latus rectum, though, a bit mysterious.In general though all we're doing is defining a conic section. Which. You know, this is the two body problem, of course it is. Adjust e to create the shape, while p determines the size. 

For circular orbits, e is zero and thus c is also zero, which makes sense. c, the Laplace Vector, has to maintain not just magnitude but also direction as it is completely conserved, and for a circular orbit the direction of r changes to all directions but nothing else does, so in order to keep the difference the same the Laplace Vector must be zero.
