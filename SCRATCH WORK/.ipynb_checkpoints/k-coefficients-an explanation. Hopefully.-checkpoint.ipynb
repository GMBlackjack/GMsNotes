### Let's explain it anyway

It all starts from this equation which is never explained very well. [1, though it is stated in several places in slightly different ways. The typography in this paper makes the expression ambiguous, but other papers confirm that the $\nu$ is a subscript of k]

$$ T(u) = \frac{1}{\Delta \nu} \int_{\Delta \nu} e^{-k_\nu u} d\nu $$

This is known as the band mean transmission. Let's be very explicit about what each term is. $T_{\Delta \nu} (u)$ is the transmission over a certain "band" of frequencies $\nu$ as a function of u. The difficulty here is determining what *u* is. 

From [2] we get the hint that for constant $k_\nu$ we have $u = \int \rho(z) dz$, that is the integral of the density along the atmospheric path. This essentially compresses the density into a 2D density, a sort of cross section, which is to say a total mass that is passed through per unit area. In short, a measure of "how much stuff" is in our way. ([3] calls it the path length though, confusion?) 

$k_\nu$ is the absorption coefficient for a given frequency $\nu$ [2]. Naturally this is implied to be "per species" since every species would have a different value for $k_\nu$. Could also be thought of as $k(\nu)$ which is why the integral is not trivial. $\Delta \nu$ is just the range of frequencies we integrate over. In theory k can be calcualted [3] but in practice it seems likely to just be tabulated for use.  

Keep in mind that the equation overall only applies when $k_\nu$ is constant over the beam path. In any large-scale real situation this would obviously not be the case, but if we work with small slices of an atmosphere the density won't change much along the path and $k_\nu$ can be treated as constant. (Though how we tell how thin we need to slice things, I still don't know.) 

Let's examine some behavior of this function. If $k_\nu$ is constant with frequency, that is we are integrating over a small band or a smooth area of the spectrum, the result is:

$$ T(u) = e^{-k_\nu u} $$

As the integral is over a constant, the normalization factor out front simply cancels everything out. In the specific case of no absorption, $k_\nu=0$, we simply get

$$ T(u) = 1 $$

Which is to say everything transmits. This also applies when u=0, that is, there is no material in the way. As the function itself is a negative exponential, whenever we actually have absorption occur, the transmission will be a fraction less than 1, but we can't get it to actually equal 0 in this continuum situation (though arguably eventually we get to the point where T is so low not even a single photon makes it through). 

One can imagine taking the limit as $\Delta \nu \rightarrow 0$ to get an exact solution, but we won't be doing that to arrive at our correlated k coefficients. 




Our next step is to prove that we can change the variable from frequency to k like so:

$$ T(u) = \int_0^\infty f(k) e^{-ku} dk $$

This is simply stated in [1] but we want a little more justificaiton for this fundamental step, because once k is the variable we can start doing things to it that give us our k-coefficients. In this case f(k) is some normalized function of k. (Normalized on the domain 0 to infinity. k doesn't go below zero so this is probably obvious). We will concern ourselves with what it actually *does* later. 

So, how is this justified? Well...

$$ \frac{1}{\Delta \nu} \int_{\Delta \nu} e^{-k_\nu u} d\nu = \int_0^\infty f(k) e^{-ku} dk $$

Logically this makes sense, f(k) is an arbitrary function that can take any form, so it can take any role we wish in making the integrals equal, including simply recreating the previous integral (and its normalization factor) and canceling the exponential term. As f(k) itself is normalized the integral can have an absoltue maximum value of 1 and no higher, and u=0 still produces 1, and other behaviors match. The question is what exactly is f(k) such that the result for transmission T is the exact same as before?

Well, in short it is the probability distribution function for k over the range $\Delta \nu$. Any given k appears either very few times along $k_\nu$ or a lot of times, and the ones that show up a lot get higher probaiblity densities, while those that show up less get small ones. This is likely best illustrated with a graph from [1]:



This is a somewhat idealized representation of the transformation taking place here. On the left we have various values of absorption coefficients for their frequencies. On the right is the transformation of their probability densities. The shaded area just points out the places of "higher absorption" for both bands, and while they are in different spots in plot b) they represent the same frequencies in plot a). 

One might conceivably ask why we can be sure integrating $f(k)e^{-ku}$ will actually give us the same T as if we did it over frequency. Well, think about it, the value for $e^{-ku}$ hasn't actually changed, so every k will still produce the correct value for the exponential, f(k) is a *weighting function* that makes sure everything is given the proper value it would have been given if we did it in terms of $\nu$. After all, The only $\nu$ inside the integral is *in* $k(\nu)$, so this *must* be possible to do this (assuming continuity). As $\Delta\nu$ is some constant it can just be folded into f(k). f(k) itself can be thought of as the "k-distribution."  

f(k) can also be integrated to get g(k), a cumulative distribution function (CDF), which can also be used. (CDFs are just another sort of distribution, where instead of "probability density at this spot" it is "probability of being before this spot."). We can then relate this back to the transmission again by inverting g(k) into k(g). [3].

$$ T(u) = \int_0^1 e^{-k(g)u} dg $$

Note that we have no normalization out front. This is because the "length" is 1, as g is a CDF. 

What is the "correlated k-distribution" then? Well, it's k(g). That is, we put in a value from 0-1, representing a value of g(k), and get the k out. (Basically the inverse function). Now this is naturally very hard to do analytically but we're here making computerized simulations so we can just manually inver the funciton. Quesiton is: why though? 

To answer this let's take a step back. Each individual k-distribution is only valid for a very small area of the atmosphere where the k itself is constant with depth. One of the things that changes with depth is pressure (usually), and we can actually see a pressure change in the graph above, the solid and dashed lines have their own presure values. They are meant to represent different parts of the same situation--and the important part is that they have *similar shapes*. This is where the *correlated* in *correlated k-coefficients* comes from, the idea that even as the situation changes, the *same frequencies will have peaks even if the lines themselves have been broadened*. From this a relation can be constructed that isn't restricted to just one physical situation, but can continue through an atmosphere. 

So, let's find that relation. Start by writing out the transmission over frequency with *variable* k. 

$$ T = \frac{1}{\Delta \nu} \int_{\Delta \nu} e^{-\int_{z_1}^{z_2} k(\nu,P,T) \rho dz} d\nu $$

The z values are just heights that define the atmosphere, or the section of an atmosphere we care about. Now the correlated k method eventually provides the following: [3].

$$ T = \int_0^1 e^{-\int_{z_1}^{z_2} k(g,P,T) \rho dz} dg $$

The math that shows this is not obvious but since the results work we can assume that, at least under most conditions we care about, the converstion to g applies here--though it is an approximation, specifically one that depends on the k values being *correlated*. Now, question is, how do we *use* this? 

Our goal is, ultimatley, to get T, the transmission. Without the correlated k-distribution method, we have to integrate over $\nu$ which has a lot of fine details. However, if we can avoid that and integrate over g instead, g is a *smooth* funciton, and thus is much easier to integrate. How much better can it be? Well, just look at this example from [3]. 


 
Thus we have an effective method if we can convert $k(\nu,P,T)$ to $k(g,P,T)$. How do we DO this though?

First, assume you HAVE a k in terms of $\nu$, P, and T. The form given in [3] is 

$$ ln k(\nu, P, T) = \sum_{n=0}^2 a_n(\nu, P) (T-250)^n $$

$a_n$ are constants that have, hopefully, been tabulated for whatever species you happen to care about. We attain g through the following:

$$ g(k,P,T) = n(0,k;P,T)/N $$

Now we know what g is, but what is n? This is the "number of wavenumber points at which the absorption coefficients k are between 0 and k", and N is the total number of points. This is, by nature, not an exact solution but a numerical one, we're sampling poitns of k and finding the number that exist in a certain set. With this, we have g and k, and then we just need to invert the relation to find k(g). Except P and T are still there, annoyingly. 

Since g is equivalent to $\nu$ and we *have* a relation between g and k, we know the following must be true:

$$ ln k(g, P, T) = \sum_{n=0}^2 A_n(g, P) (T-250)^n $$

Notably there are other ways to do this. [2] goes from k to f(k) then to g(k) and then inverts it to k(g), for instance. The main point is that integrating over $\nu$ is pain while integrating over g is bliss, fewer points can be sampled for greater accuracy.  

We wanted to try to show how we can extract these simpler functions, but turns out getting probability distribution functions analytically is a pain. The *cumulative* distribution functions, however, are actually much simpler when you have analytic forms.

Basically, if you have a random variable x that is uniformly distributed, and a relation y = f(x) and you want the probability distribution for y, you use the following:

$$ PDF(y) = \frac{1}{b-a} \frac{df^{-1}(y)}{dy}$$

That inverse makes it so only one-to-one functions play nice, though. (Source? This is basic probaility stuff, but it was annoying to find, I actually used chatGPT and then confirmed by graphing and integrating the probabilities. This is how to use AI: do not trust the answer, verify it every time.) 

This was derived by finding the CDF. The CDF is a bit simpler; we know the CDF of a uniform distribution.

$$ CDF(x) = \frac{1}{b-a} x - \frac{a}{b-a} $$

Note: this is just on the range a<x<b, we have 0 on the left of that and 1 on the right. Usually though the math works by just looking at the middle. 

$$ CDF(y) = CDF(x)(f^{-1}(y)) $$

However even though this may be simpler in a sense, it still requires the *inverse* function, which again doesn't play well with humps. So INSTEAD why don't we PROGRAM A THING that will just use a monte-carlo approach to grab the new function?

```python
import math
import random

def funcEval(x):
    return math.sin(x)
    # This is the function f(x), that is, it gives us y. We want the PDF of y in the end. 

# declare our range of values. 
lowerBound = 0.0
upperBound = 4.0
# these correlate to a and b.

# how many times to we want to randomly evaluate?
evaluations = 10000

# how many bins do you want to sort your y values into at the end? 
# this determines the "resolution" of the calculated PDF.
bins = 100

#This array will store our y points for later analysis. 
yValues = []

i = 1
while (i <= evaluations):
    #This loop evaluates the function "evaluations" times and records all found y values.
    yValues.append(funcEval(lowerBound + random.random()*(upperBound - lowerBound)))
    i = i+1

print(yValues)

# Now what remains: sort y values into bins. 
# Bin values need to be normalized so the addition of all bins equals 1.
# then plot with matplotlib. Consider result. Make lots of plots and compare always.
```
