# PARV IV: Linear Algebra and Cramer's Rule

https://youtu.be/dnzUgDl43rQ?si=UK4fLzCOq1MvxAbA

What else can we get from this? 

Well say we are solving a linear combination equation.

$$ a_1x + b_1y = c_1 $$
$$ a_2x + b_2y = c_2 $$

Which we can write with vectors as 

$$ x\vec a + y \vec b = \vec c $$

So, how to we solve for x and y then with geometric algebra? Well, we wedge it.

$$ x\vec a \wedge \vec b + y \vec b \wedge \vec b = \vec c \wedge \vec b $$
$$ \Rightarrow x\vec a \wedge \vec b = \vec c \wedge \vec b $$

Of course solving for x here appears to need an inverse... oh or we can just divide (assuming the result isn't zero).

$$ \Rightarrow x = \frac{\vec c \wedge \vec b}{\vec a \wedge \vec b} $$

We can do this since the division is over a bivector, so really what we end up with is the "area ratio" between the two, leaving a sclaar. Also known as UNITS WORK OUT! 

Similarly for y we get $y = \frac{\vec a \wedge \vec c}{\vec a \wedge \vec b}$

This is all entirely arbitrary. Neat!

Let's work out an arbitrary wedge product... but do it from the matrix form of linear combinations so we can find something new out.

$$ A = \begin{bmatrix} 
a_1 & b_1 \\
a_2 & b_2 \end{bmatrix} $$

Let's do the wedge product between a and b.

$$ \vec a \wedge \vec b = (a_1e_1 + a_2e_2)\wedge(b_1e_1 + b_2e_2) = a_1b_2(e_1\wedge e_2) + a_2b_1(e_2 \wedge e_1)$$
$$ = (a_1b_2 - b_1a_2) (e_1 \wedge e_2) = det[A] e_1e_2 $$

This is the familiar 2x2 matrix determinant. Which is usually just some mysterious number but now we have a way to derive it. (But what of higher order determinants?) 

If we replace wedge products with determinants we get Cramer's rule. In short cramer's rule is

$$ x_i = \frac{det[A_i]}{det[A]} $$

The specifics of how this work is irrelevant to our purposes, as it involves some linear algebra. Our victory here is that we can get the same result from geometric albegra without finagling with the normal proof for Cramer's rule. 

A note on the determinant = 0 case. Since the determinant is an outer product, and an outer product is the paralellogram area, that implies that a det=0 case would be between two vectors that are parallel and thus have no wedge product. Visualization power, activate!

The situation, as we have it, is gravity. We have some relations here:

$$ \vec F_{12} = G \frac{m_1m_2}{r^2}\hat r $$

And thus

$$ \ddot r_1 = G \frac{m_2}{r^2}\hat r $$
$$ \ddot r_2 = - G \frac{m_1}{r^2}\hat r $$ 

And also by taking derivatives

$$ \ddot r = \ddot r_2 - \ddot r_1 $$
$$ \Rightarrow \ddot r = - \frac{G(m_1+m_2)}{r^2} \hat r $$
$$ \Rightarrow \ddot r = - \frac{\mu}{r^2} \hat r $$

Where $\mu$ is just there for bookkeeping and is known as the gravitational parameter. 

There has been no geometric algebra yet. This is just the radial component of our problem.

We'll also be needing the angular momentum:

$$ \vec L = \vec r \times \vec p $$

However the cross product is really annoying and is inherently three dimensional. What if we really do wish to restrict ourselves to 2D? Well, we can do that! With GEOMETRIC ALGEBRA! You can probably see how one might represent the cross product using bivectors if you are familiar with the cross product. Otherwise just strap in and let's go!

So, define our coordinates. Let $\vec r$ be the vector from the pivot, and $\dot r$ be the direction of motion. We can create a bivector from this:

$$ H = \vec r \wedge \dot r $$. Naturally this does, in fact, circulate in the direction of rotation. For circular motion it's also easy to see that nothing particuarly changes here. However, now we need to take the derivative of a wedge product. One might think it odd but the product rule still applies here. 

$$ \dot H = \dot r \wedge \dot r + \vec r \wedge \ddot r = \vec r \wedge \ddot r $$ 

Now, remember what we showed in the previous part about $\ddot r$ pointing in the -r direction? Well paralell wedge products are zero. This means..

$$ \dot H = 0 = \frac{d}{dt} (\vec r \wedge \dot r) $$

This "the time derivative is zero" is the definition of a CONSERVED QUANTITY! Man that was easy. 

Now let's move to Kepler's Second Law. (don't think too hard about the order). 

So, Kepler's Second Law is about the area swept out at any given moment by an orbiting body. Well, we can represent the motion of an object with the position vector and tis velocity vector, so we can construct a paralellogram $\vec r \wedge \dot r dt$. Of course the actual area swept out is a triangle so we divide thsi by two. Thus:

$$ dA = \frac12 \vec r \wedge \dot r dt $$
$$ \dot A = \frac12 \vec r \wedge \dot r $$
$$ \dot A = \frac12 H$$

So a nice result. However we do need to point out that we pulled some weird derivative jujitsu there. Careful when dividing by differentials. 

The other point of this result is that since H is conserved, that means the rate of area swept out is constant. Which is Kepler's Second Law. Yay!

Okay so now let's look at some weird unusual things. First of all, show that $\vec r \cdot \dot r$ is zero. We do this by pulling *out* a time derivative, doing the product rule in reverse.

$$ = \frac12 \frac{d}{dt}(\vec r \cdot \vec r) $$

But wait, a dot product... is a constant. And a constant has a time derivative of zero. Thus, the dot product of a vector with its "velocity" vector is zero. This would appear to be absolutely general. (The video uses unit vectors for this but changing the magnitude will have absolutely no effect on this). 

Anyway, that's useful, and so let's use it to figure out more about H.

$$ H = \vec r \wedge \dot r = (r \hat r) \wedge (\dot r \hat r + r \frac{d}{dt} \hat r) $$

I will point out that I would love to put a derivative, hat, and vector on multiple things, but LaTeX wont' let me. So we now have notation confusion abounding. Lovely. Anyway, regardless of all that, what we've done is split the components up into their individual parts, reminding everyone that $\dot r$ as a vector would actually have multiple components and be subject to the product rule. Anyway distribute through the wedge product now--the $\hat r$ terms cancel due to being the same and we are left with

$$ = r\hat r \wedge r \frac{d}{dt} \hat r = r^2 \hat r \wedge \frac{d}{dt} \hat r = r^2 (\hat r \frac{d}{dt} \hat r - \hat r \cdot \frac{d}{dt} \hat r )$$

We previously showed that the dot product goes to nothing. And so...

$$ H = r^2 \hat r \partial_t \hat r $$

Where we introduce the relativistic version of the time derivative operator to keep us all sane. For the record this is the angular momentum per unit mass equation.

This should be kind of obvious as the exponentials will combine on either side and the signs will be opposite but magnitudes will be equal. So yeah. 

$$ e^{-\frac12 \theta_1 B_1 - \frac12 \theta_2 B_2} u e^{\frac12 \theta_1 B_1 + \frac12 \theta_2 B_2} $$

Since each sum of bivectors is another bivector, that means there is a single bivector with a single constant in front of it, so we can always rework this into a single rotor. Tah-dah! 

Bizarre property: $RuR^\dagger = (-R)u(-R^\dagger)$. Which basically states abstractly that you can rotate backwards to get the same result if you want. Mathematically this basically states that you always have two rotor sets that can do one rotation, they are not unique. 

Very unusual property about rotors. If you apply a rotor TO a rotor, you get another rotor. But the rotation that should cause no change *doesn't*. Take, for instance, $R_1uR_1^\dagger$. Let's rotate this again by $R_2$, where $R_2$ *specifically* is the rotation of a full circle, $2\pi$. This means $R_2 = e^{\pi B}$. However, when you alter an exponential by the angle $\pi$, you don't fully rotat it, you flip its sign. Thus,

$$ R_2R_1uR_1^\dagger R_2^\dagger = (-R_1) u (-R_1^\dagger) $$

So the final result is still the same, but the rotors themselves flip sign when rotated a full circle. It takes $4\pi$ to get them back to where they started. This seems unusual and esoteric and just a weirdness of the calculation that goes away when you carry it out. If that was all it was I wouldn't be writing it down, because there's quantum spin weirdness that actually needs to rotate twice to return to where it was. It can actually be used to model things. They are called SPINORS. (both in physics and as well as pure math as an object that has to spin twice to return to its origin).

# PART XIV: Linear and Spherical Interpolation

https://youtu.be/ibkT5ao8kGY?si=_hfXk8JrCGJYsMjO



So a new application arises. Interpolation.

So interpolation is a bit like this: take two vectors. Draw a line between their two tips. We want a vector that points somewhere along that line, the "Somewhere" defined by a quantity s that scales from 0 to 1.



$$ v_s = v_i + s(v_f-v_i) = (1-s)v_i + sv_f$$

Rather simple really.

Now let's move to Spherical Interpolation. A.K.A. SLERP. I love silly acronyms. 



The main things of interest here are that we're not interpolating over a line, we're interpolating over an arc. Specifically this arc has MANY great properties: it's part of a great circle, it's the shortest path between the two points of the vectors (well if the vectors really are on the unit sphere as they shoudl be), it's also directly in line with the bivector plane created by the two vectors, it's just full of great stuff. This means that the spherical interpolations are *rotations*. 

Which leads to the somewhat obvious thought of "well in that case let's just use the rotors to do it." We'll see if that's where the video goes. 

Turns out that yes we can do it. Since the rotation is in a 2D sense we can just do it with one side, but in general we want to use a rotor sandwich. Just scale the angle and make sure we know the plane B, BUT since we are on the unit sphere in the first place we already have two vectors that can be used to defien B. Just make sure we have a full definition. Just replace all $\theta$ by $s\theta$. Tah-dah.

Also can be represented as $R^s$. s only goes from 0 to 1, and the exponential absorbs powers as products in the power $(e^x)^s = e^{sx}$, easy. 

Now, what if we want a direct formula? Well, that's a little different. 

$$ v_s = av_i + bv_f $$

We seek expressions for a and b. Using the single-rotor notation (which we can do in 2D situations) we simplify it greatly. 

$$ \Rightarrow v_i R^s = av_i + bv_iR $$
$$ \Rightarrow v_i R^s \wedge v_iR = (av_i + bv_iR)\wedge v_iR $$
$$ \Rightarrow \frac12(v_iR^sv_iR - v_iRv_iR^s) = \frac12 a(v_i^2R - v_i R v_i)$$

Note that the b-term vanished here as we wedged the vector with itself. 

$$ \Rightarrow v_i^2R^{s\dagger}R - v_i^2R^\dagger R^s = a(v_i^2R - v_i^2 R^\dagger)$$
$$ \Rightarrow R^{s\dagger}R - R^\dagger R^s = a(R - R^\dagger)$$

The unit vectors are gone as they all squared, and now we have just a bunch of rotors. 

$$ \Rightarrow R^{s\dagger}R - (R^{s\dagger}R)^\dagger = a(R - R^\dagger)$$

A rotor minus its conjugate and a rotor minus its inverse on both sides, scaled by a constant. Curious. A property of this type of inverse, the conjugate, is that the "imaginary" component is all that remains when you subtract the two--in this case it would be only the bivector component that remains. 

$$ \Rightarrow 2 sin(\theta(1-s)) B = 2 a sin\theta B$$
$$ \Rightarrow sin(\theta(1-s)) = a sin\theta$$
$$ \Rightarrow \frac{sin(\theta(1-s))}{sin\theta} = a$$

Which is a closed expression for a. Note that we could have gotten to this via imaginary numbers... but that would be rather artificial, here it just arises from the geometric algebra. 

Similar arguments exist for getting b.

$$ b = \frac{sin(s\theta)}{sin\theta} $$

So we have

$$ v_s = \frac{sin(\theta(1-s))}{sin\theta}v_i + \frac{sin(s\theta)}{sin\theta}v_f $$

As the complete solution. While this may have looked complicated, how would you have gone about doing this without geometric algebra rotations? Rotations without geometric algebra is quite a pain, remember, and introducing imaginary numbers is a big of a leap.

There is one thought. How *do* you find the angle between two vectors?

$$ \theta = arccos(v_i \cdot v_f) $$

This *should* be simple if you have the compositions of the angles. 
 
In the small angle approximation:

$$ v_s = (1-s)v_i + sv_f $$

We recover linear interpolation.

((By the way the dagger operator is "reverse the order" operation, so ab -> ba or abc -> cba)). 

Final topic: NORMALIZED linear interpolation! 

WHich is to say you take the linear interpolation and normalize the result. Not required for SLERP since the unit sphere is automatically considered to be present. Gotta say, not much special about this, but do note that it won't produce the same result as SLERP since angles may not match up with linear scaling.

Why do we care? Probably application dependent.

Hmm those are the unit bivectors. Neat. In essence I maps the vector to the bivector it is normal to, and it follows the right hand rule for circulation. We can say this makes the vectors and bivectors "Dual" to each other. I is a duality transformation here. And in 3D we can go between vectors and bivectors directly.

In 3D, we have the situation where something of grade r when acted upon by I becomes 3-r. so scalars->pseudoscalar, vectors->bivectors in 3D. In other dimensions this still applies apparently, though it does mean bivectors would map to themselves.

Cross product anti commutes.
