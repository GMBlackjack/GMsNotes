# New Mask Editor Stuff

## For editing the mask

### The new mask made specifically of Titan.

Just placing code here that'll take the image of the mask, display it, and then edit it.



```python
%matplotlib ipympl
# READ IN FILES

#IMPORTS 
import csv
import math
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image

#Now we need to read in the images so we can manipulate them.
im1 = Image.open("1DegreeMap.tif")
ima1 = np.array(im1)

fig, ax = plt.subplots(figsize = (8,8))
image = ax.imshow(ima1)
#Tifs are stored by y, x, and then RGB. RGB is mostly irrelevant to us though, as this is monochrome. 
```

```python
print(len(ima1))
print(len(ima1[0]))
print(len(ima1[0][0]))
```

```python
%matplotlib ipympl

x,y = 0,0
test = 0 #Point is unknown

while (y < len(ima1)):
    x=0
    while (x < len(ima1[0])):
        #Basically, only if not out of bounds do we do anything. 
        test = 0
        if (ima1[y][x][0] == 0):
            if (ima1[y][x][1] == 0):
                if (ima1[y][x][2] == 0):
                    test = 1 #Point is good!
                else: 
                    test = -1 #Point is bad!
            elif (ima1[y][x][1] == 255):
                if (ima1[y][x][2] == 255):
                    test = 1
                elif (ima1[y][x][2] == 0):
                    test = 1
                else:
                    test = -1
            else:
                test = -1
        elif (ima1[y][x][0] == 255):
            if (ima1[y][x][1] == 0):
                if (ima1[y][x][2] == 0):
                    test = 1
                elif (ima1[y][x][2] == 255):
                    test = 1
                else:
                    test = -1
            else:
                test = -1
        else:
            test = -1
        if (test == -1):
            ima1[y][x][0] = 255
            ima1[y][x][1] = 255
            ima1[y][x][2] = 255
        x=x+1
    y=y+1

plt.close()
fig, ax = plt.subplots(figsize = (8,8))
image = ax.imshow(ima1)

Image.fromarray(ima1).save("NewTitanMask.bmp")
```
