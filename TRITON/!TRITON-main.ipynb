So, Triton is exacessively weird even from a distance. Even if you know nothing about it besides the fact that it orbits Neptune, that alone directs to a very important fact:

Triton. Orbits. *Retrograde*.***BACKWARDS.***



*Triton orbit and inclination compared to Neptune. From https://commons.wikimedia.org/wiki/File:Triton_orbit_%26_Neptune.png*

No other major moon in the solar system does this, only small moons do, and it's basically accepted that you need to capture moons to get them in retrograde (or something even crazier has to happen, one could imagine a moon-on-moon collusion occuring to get some of the small ones in their weird orbits). Thus, despite it sounding crazy, it is generally accepted that Triton is a captured satellite and that it used to be a Kuiper Belt Dwarf Planet like Pluto. 

This is reinforced by the fact that Triton is made of essentially the same stuff as Pluto, has a similar size, and a similar density. (Triton is larger. In fact it's larger than all the dwarf planets which is a little suspicious... why did Neptune happen to catch the biggest one?) 

Anyway, due to tidal forces, Triton's retrograde orbit means that it's slowly inspiraling to Neptune and is eventually going to be torn apart to form a new ring system [1]. When? 3.6 billion years [1]. It's got a while yet. 

Let's also talk about the orbit itself. Titan, in addition to orbiting retrograde, is also orbiting at an extreme inclination angle--yet another reason to think it didn't form in Neptune's primary disk. This angle gives Titan all sorts of fun seasonal effects which we'll get into later. 

But here's another thing. The orbit? Almost perfectly circular, eccentricity is so low you write it as 0.00 most of the time. Written out, we have ~1e-5 eccentricity [2]. Earth's eccentricity is, for the record, 0.02. Venus is the most circular of the planets, with 0.007. Triton's by far the roundest of the major moons, though there are some tiny moons with also very circular orbits. Suddenly this freaky moon's orbit is far more regularized than is natural.

Triton has a tenuous atmosphere that nonetheless has sufficient winds to perturb its surface. 

Atmosphere is nitrogen dominated. Has traces of methane and carbon dioxide, as it is warm enough to partially sublime the ices. (Though there is VERY little CO2, there's more hydrogen than it). 

The atmospheric pressure was 14$\pm$1 ubar in 1989. Formed by sublimation and, maybe, geyser activity. Yes there is known geyser activity. Pressure was shown to increase by 5 um in the 1990s. This change means either a high thermal inertia for Triton due to the atmosphere increase continuing after favorable sun geometries, or changes in the albedo and emissivity are driven by the seasonal cycle. 

Triton experiences condensation flows where the gas condenses at the winter pole and sublimes at the summer pole. This may be why ice overlays Triton's equatorial regions (the cold ones) while there is no ice in the warmer areas. 

Just, Geysers. Let's address the fact that geysers exist. 

Due to the heavy inclination Titan undergoes seasons within seasons on a scale of about 600 years. 

Triton is far larger than any other Neptunian moon.  

Most of Triton is covered in nitrogen/methane ice, but it can't hide the terrain entirely. The western (Trailing) hemisphere has the catalope terrain. This is the oldest terrain on Titan and yet having minimal craters it must be geologically young. 

The leading hemisphere has a smoother surface with large calderas/lava lakes, probably a few billion years old. (geolocially "young" eh?) 

Ice on the two hemispheres is different, with different ice viscocities, indicating a different composition. Ammonia on one side? 

Ice lava lakes show cooling and stagnation levels. Though it is a bit mysterious as we do not know if the features are formed by drainage of lava lakes or collapsed calderas. 

The polar nitrogen ice has a slight reddish tint, indicative of organics. Of wht kind? we've been here before we don't know.

Dark albedo areas are found at the poles, two of which seem to be indicative of active geysers. The plumes rise 8 km and can extend for 100 km. All known geysers (four) are at the south pole, which was the summer pole at the time. Theory: solid-state greenhouse effect leading to increased subsurface nitrogen pressure?

Tidal heating was probably important in Titan's past. 

Triton's density is very similar to Pluto, and it's more rocky than the other large gas giant moons (excepting Io and Europa, which kind of have their own thing going on). 

One might expect Triton to provide hydrogen ions to Neptune's magnetosphere, but the values are far lower than expected, though Triton does have a loose hydrogen cl9oud. (And hey wait, actual clouds too, why didn't the book mention that!?)
